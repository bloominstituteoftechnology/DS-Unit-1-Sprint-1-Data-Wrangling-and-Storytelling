```python
# Your Work Here # Code is in lesson notebook, hopefully I've uploaded it.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from IPython.display import display, Image

url = 'https://fivethirtyeight.com/wp-content/uploads/2017/09/mehtahickey-inconvenient-0830-1.png'
example = Image(url=url, width=400)

display(example)
```

```python
fake = pd.Series([38, 3, 2, 1, 2, 4, 6, 5, 5, 33], 
                 index=range(1,11))

fake.plot.bar(color='C1', width=0.9);
```

```python
import matploblibrc
```

```python
# Set Stylesheet
plt.style.use('fivethirtyeight')

# Figure axis syntax
fig, ax = plt.subplots()

# Make figure background color white
fig.patch.set(facecolor='white')

# put bars on the axes
fake.plot.bar(width = 0.9, color='#ED713A')

# set axes background color
ax.set(facecolor='white')


ax.set_title("'An Inconvenient Sequel: Truth To Power' is divisive", loc='center', fontsize=12, fontweight='demibold', x=.35, y=1.1)
ax.set_ylabel('Percent of total votes', fontsize=10, fontweight='bold')
ax.set_xlabel('Rating', fontsize=10, fontweight='bold')
#Subtitle
ax.text(s='IMDB ratings for the film as of Aug. 29th', y=42, x=-1.88, fontsize=11)

# Set Ytick Labels
ax.set_yticklabels(['0', '10', '20', '30', '40%'], fontsize = 10, color='lightgray', fontweight='demibold')

# Set Yticks (fixed grid lines too)
ax.set_yticks(range(0, 41, 10))

# Set X tick labels
ax.set_xticklabels(range(1,11,1), rotation='horizontal', fontsize=8, fontweight='bold', c='lightgray')

# Add bottom text / line
ax.text(s='FiveThirtyEight', y=-11, x=-1.7, fontsize=7.3)
ax.text(s='SOURCE: IMDB', y=-11, x=8.1, fontsize=7.3)
ax.text(s='------------------------------------------------------------------------------------',y=-10, x=-1.65, color='lightgrey')
fig.show()
```
