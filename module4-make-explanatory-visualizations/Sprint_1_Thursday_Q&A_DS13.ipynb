```
df = pd.read_csv('ncaa-womens-basketball-tournament-history.csv')

print(df.shape)
df.head()
```

```
df['Year'].value_counts()
```

```
Louisiana = df[df['School'] == 'Louisiana Tech']

Louisiana.head()
```

```
Louisiana['Year'].value_counts().sort_index()
```

```
Louisiana = Louisiana.reset_index()

Louisiana.head()
```

```
Louisiana['Year']
```

```
Louisiana['Seed']
```

```
plt.scatter(Louisiana['Year'], Louisiana['Seed'])
plt.plot(Louisiana['Year'], Louisiana['Seed'])

plt.show()
```

```
Louisiana.dtypes
```

```
Louisiana['Seed'][0]
```

```
Louisiana['Seed'] = Louisiana['Seed'].astype('int')


Louisiana.dtypes
```

```
def one_hundred_minus_seed(cell_contents):
  return 101 - cell_contents

Louisiana['100_minus_seed'] = Louisiana['Seed'].apply(one_hundred_minus_seed)

Louisiana.head(15)
```

```
# plt.scatter(Louisiana['Year'], Louisiana['100_minus_seed'])
plt.plot(Louisiana['Year'], Louisiana['100_minus_seed'])

plt.show()
```

```
import scipy

fig, ax = plt.subplots()

# Line PLot
ax.plot(Louisiana['Year'], Louisiana['100_minus_seed'])

# Set the title
ax.set_title("The fall of yesterday's women's basketball powerhouses", fontsize=14, fontweight='bold', y=1.1)

# Set a subtitle
ax.text(s="Seed points by year for Louisiana Tech", x=1975 , y=110, fontsize=12)

ax.set_xlabel("Year", fontsize=12, fontweight='bold')

ax.set_ylim(80,102)

# Fill in below the line
d = scipy.zeros(len(Louisiana['100_minus_seed']))
ax.fill_between(Louisiana['Year'], Louisiana['100_minus_seed'], where=Louisiana['100_minus_seed']>=d, interpolate=True, color='blue')

# ax.fill_between(xs, ys, where=ys<=d, interpolate=True, color='red')
plt.show()
```
