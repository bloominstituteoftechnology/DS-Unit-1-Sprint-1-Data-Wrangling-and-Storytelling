```python
pepe = "https://raw.githubusercontent.com/fivethirtyeight/data/master/rare-pepes/pepecash_prices.csv"

df2 = pd.read_csv(pepe)

df2
```

```python
df2.isnull().sum()
```

```python
df2['Timestamp'] = pd.to_datetime(df2['Timestamp'], unit='s')

df2

```

```python
df2['year'] = df2["Timestamp"].dt.year
df2['month'] = df2["Timestamp"].dt.month

df2['year'] = df2['year'].astype(str)
df2['month'] = df2['month'].astype(str)

```

```python
df2['Month_Year'] = df2['month'] + '/' + df2['year']

df2
```

```python
df2 = df2.drop(['Timestamp', 'year', 'month'], axis=1)

df2
```

```python
df2["Month_Year"].value_counts().sort_index()
```

```python
df2['Month_Year'].dtype
```

```python
filter = df2[(df2['Month_Year'] > '2017')]
filter.head()
```

```python
#quick and dirty first

plt.plot(df2['Month_Year'], df2['Price'])
```

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=((7,3)))

ax.plot(filter['Month_Year'], filter['Price'])

fig.patch.set(facecolor='white')
ax.text(s="The Changing Value of Pepe Cash", y=1, x=1, fontsize=15)
ax.text(s="Price of Pepe Cash in U.S. Dollars", y=0, x=1, fontsize=15)

ax.set (facecolor='white')

ax.set_xlabel("date")

ax.set_ylabel("price")

plt.show()

```
