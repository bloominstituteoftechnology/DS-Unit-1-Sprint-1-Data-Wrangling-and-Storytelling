```python
# This is my desired graph. It's a 3d color map of heat transfer calculating the thermal analysis. I made
# a few adjustments to the font settings and to the coloring.

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Note: Had to import 'Axes3D' from 'mpl_toolkits.mplot3d' in order for the 3D chart to function

n, p = 10, .5
s = np.random.binomial(n, p, 1000)

fig = plt.figure()
ax = fig.gca(projection='3d')

X = np.arange(-3, 5, 0.25)
Y = np.arange(-3, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

surf = ax.plot_surface(X, Y, Z, cmap=cm.viridis,
                       linewidth=0, antialiased=False)

ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Personalized font types to use on the charts.
font = {'family': 'serif',
        'color':  'teal',
        'weight': 'bold',
        'size': 10,
        }
font2 = {'family': 'serif',
        'color':  'cyan',
        'weight': 'regular',
        'size': 9,
        }

ax.set_title("Heat Transfer Profile", fontsize=11, verticalalignment='top', fontdict=font)
ax.set_xlabel("Width", verticalalignment='center', fontsize=11)
ax.set_ylabel("Length", fontsize=11)
ax.set_zlabel("Thermal Analysis", fontsize=10, verticalalignment='bottom', fontdict=font2)
ax.set_xticklabels(labels=['', '-2', '0', '2', '4'], fontsize=10)
ax.set_yticklabels(labels=['', '-2', '0', '2', '4'], fontsize=10)

# The code '.labelpad' below is to adjust the position of the x-axis and z-axis labels.
ax.xaxis.labelpad = 1
ax.zaxis.labelpad = 6.5

# This is to adjust the color bar on the right hand side.
fig.colorbar(surf, shrink=0.7, aspect=6)

# Finally, we're going to set the background color of both the figure and the axes.
fig.patch.set(color='gray')
ax.set_facecolor(color='silver')

plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

ratings = pd.Series([38, 3, 2, 1, 2, 4, 6, 5, 5, 33], 
                 index=range(1,11))

plt.style.use('fivethirtyeight')

# First, we're going to set the axes syntax.
fig, ax = plt.subplots()

# Now we're going to add the function for the bar chart.
ratings.plot.bar(ratings, width=0.9, color='#39ed99')

# Now we're going to set the title. I added my personalized 'fontdict=' from the 3d chart above.
ax.set_title("'An Inconvenient Sequel: Truth To Power' is divisive", 
             loc='left', 
             color="cyan", 
             fontsize=12, 
             fontweight='roman',
             fontdict=font,         
             x=-.07, 
             y=1.1)

# Next we're going to do practically the same thing as with the title, but with a different 'fontdict='
# and of course a different position of the axis.
ax.text(s='IMDb ratings for the film as of Aug. 29', 
        color='teal', 
        fontweight='light',
        fontsize=11,
        fontdict=font2,
        x=-1.4,
        y=42)

# This is the customized x-label.
ax.set_xlabel('Rating', color='cyan', fontsize=11, fontweight='semibold', fontdict=font2)

# This is the customized y-label.
ax.set_ylabel('Percent of total votes', color='cyan', fontsize=11, fontweight='semibold', verticalalignment='top', fontdict=font2)

# This is the customized tick labels of the x-axis.
ax.set_xticklabels(labels=range(1,11,1), color='teal', rotation='horizontal', fontsize='11', fontdict=font2)

# This is the customized tick labels of the y-axis.
ax.set_yticklabels(labels=["0", "10", "20", "30", "40%"], color='teal', fontsize='11', fontdict=font2)

# This is to adjust the position of the y-axis label.
ax.yaxis.labelpad = 3.5

# This is to set the range of the y-ticks.
ax.set_yticks(ticks=range(0,50,10))

# This is to set the background color.
fig.patch.set(color="black")

# This is to set the background color of the axes.
ax.set_facecolor(color='teal')

plt.show()
```

```python
import altair as alt
from vega_datasets import data

source = data.unemployment_across_industries.url

selection = alt.selection_multi(fields=['series'], bind='legend')

alt.Chart(source).mark_area().encode(
    alt.X('yearmonth(date):T', axis=alt.Axis(domain=False, format='%Y', tickSize=0)),
    alt.Y('sum(count):Q', stack='center', axis=None),
    alt.Color('series:N', scale=alt.Scale(scheme='category20b')),
    opacity=alt.condition(selection, alt.value(1), alt.value(0.2))
).add_selection(
    selection)
```
