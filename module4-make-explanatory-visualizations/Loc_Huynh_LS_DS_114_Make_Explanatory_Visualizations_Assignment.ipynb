```python
# Set stylesheet
plt.style.use('fivethirtyeight')

# figure axes syntax
fig, ax = plt.subplots()

# Make the figure background color white
fig.patch.set(facecolor='white')

# Put bars on the axes
mygraph.plot.bar(width=0.9, color='#ED713A')

# Set axes background color
ax.set(facecolor='white')

#Set plot title
ax.set_title("'An Inconvenient Sequel: Truth To Power' is divisive", fontsize=12, fontweight='bold', x=.35, y=1.1)

# Set plot ylabel
ax.set_ylabel('Percent of total votes', fontsize=9, fontweight='bold')

# Set plot xlabel
ax.set_xlabel('Rating', fontsize=9, fontweight='bold')

# Set subtitle
ax.text(s='IMDb ratings for the film as of Aug. 29', y=41.5, x=-1.9, fontsize=11)

# Set y axis labels, tick marks and grid lines
ax.set_yticklabels(['0', '10', '20', '30', '40%'], fontsize=10, color='grey')
ax.set_yticks(range(0,50,10))

# Set x axis labels, tick marks and grid lines
ax.set_xticklabels(range(1,11,1), rotation='horizontal', fontsize=10, color='grey')

ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# Show the plot and squelch the funny output
fig.show()
```

```python
plt.style.use('fivethirtyeight')

fig, ax = plt.subplots()

newgraph.plot.bar(width=0.9, color='C3')

# Set plot title
ax.set_title('Comics Are Gaining A Few LGBT Characters', fontsize=15, fontweight='bold', x=.40, y=1.2)

# Set subtitle
ax.text(s='LGBT characters introduced in to DC and Marvel comics per year,', x=-6.40, y=17.5, fontsize=11)
ax.text(s='including retroactive continuity changes', x=-6.40, y=16.5, fontsize=11)     

# Set y axis tick marks and grid lines
ax.set_yticklabels(['0', '5', '10', '15'], fontsize=10)
ax.set_yticks(range(0,20,5))

ax.set_xticklabels(["1940", "'50", "'60", "'70", "'80", "'90", "2000", "'10"], rotation='horizontal', fontsize=10)
ax.set_xticks(range(5,80,10))

ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(True)
fig.show()      
```
