```python
# Your Work Here
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
'''
That line is only for jupyter notebooks, and allows plt figures to show up in your noteboook.
if you are using another editor, you'll use: 
 plt.show() 
at the end of all your plotting commands to have the figure pop up in another window.
'''
```

```python
# plot type
plt.style.use('classic') 
```

```python
# Read in the titanic dataset
url = 'https://raw.githubusercontent.com/austinlasseter/pandas_visualization/master/data/titanic.csv'
df = pd.read_csv(url)
df.head()
```

```python
# cabin class is a categorical variable
# it can be displayed using either number or string values.
df['cabin_class'] = df['Pclass'].map({1:'first', 2: 'second', 3:'third'})
df.head()
```

```python
# Fare is a continuous variable, so we can display it with a histogram.
df['Fare'].plot(kind='hist');
```

```python
# Remove two extreme outliers
df=df[df['Fare']<500]
df['Fare'].plot(kind='hist',color='dodgerblue');
```

```python
# A continuous variable can be converted to a categorical var using binning.
farebins=[0,8,15,33,512]
farelabels=['low', 'medium', 'high','vhigh']
df['fare_groups']= pd.cut(df['Fare'], bins=farebins, labels=farelabels)
myresults = df['fare_groups'].value_counts().sort_index()
myresults
```

```python
# sometimes it's important to specify the range of your y-axis to get more accurate results. 'lim'='limits'
myresults.plot();
# myresults.plot(ylim=(0,200));
```

```python
# Histogram
df['Age'].plot(kind='hist', bins=25, figsize=(16,4), color='tan');
```

```python
# Stacked Histogram of continuous variable, broken out by categorical variable
df[df['Pclass']==1]['Age'].plot(kind='hist', bins=25, figsize=(16,4), alpha=0.5)
df[df['Pclass']==2]['Age'].plot(kind='hist', bins=25, figsize=(16,4), alpha=0.5)
df[df['Pclass']==3]['Age'].plot(kind='hist', bins=25, figsize=(16,4), alpha=0.5);
```

```python
# But this is actually easier to read using a Kernel Density Estimate (KDE). Note the use of 'xlim'.
df[df['Pclass']==1]['Age'].plot(kind='density', xlim=(-5, 85), figsize=(16,4), color='navy')
df[df['Pclass']==2]['Age'].plot(kind='density', xlim=(-5, 85), figsize=(16,4), color='lightblue')
df[df['Pclass']==3]['Age'].plot(kind='density', xlim=(-5, 85), figsize=(16,4), color='crimson');
```

```python
#Displaying single categorical variable
# Bar chart. Use colors to make a particular value 'pop'!
df['fare_groups'].value_counts().plot(kind='bar', color=['gray', 'gray', 'crimson', 'gray']);
```

```python
# Pie chart
df['fare_groups'].value_counts().plot(kind='pie', figsize=(4,4));
```

```python
# Compare two categorical variables using crosstab
cab_death = pd.crosstab(df['cabin_class'],df['Survived'])
cab_death
```

```python
# Make that into a bar chart.
cab_death.plot(kind='bar');
```

```python
# Crosstab with column percents.
cab_death1 = pd.crosstab(df['cabin_class'],df['Survived'], normalize='columns')
cab_death1
```

```python
# make a stacked bar chart, for easy comparison of two categorical variables.
cab_death1.plot(kind='bar', stacked=True, color=['orange', 'gray']);
```

```python
# Crosstab with column percents.
cab_death2 = pd.crosstab(df['cabin_class'],df['Survived'], normalize='index')
cab_death2
```

```python
# A paired barchart for comparison of two categorical variables. Note the use of 'barh'.
cab_death2.plot(kind='barh', color=['purple', 'lightgreen'], figsize=(10,4));
```

```python
# Comparing a categorical and cont var
# Let's look at a single continuous variable
df['Age'].mean()
```

```python
# Now explore that continuous variable broken out by a categorical (i.e., categorical) variable.
cab_age = df.groupby('cabin_class')['Age'].mean()
cab_age
```

```python
# To visualize a single continuous variable, you can just use `.plot` with `kind=box`
df['Age'].plot(kind='box');
```

```python
# But to break out a continuous variable by a categorical variable, use `.boxplot`
df.boxplot(column='Age', by='cabin_class');
```

```python
# Comparing 2-3 cont var
# Two continuous variables: age by fare
df.plot(kind = 'scatter', x='Fare', y='Age', figsize=(12,4));
```

```python
# Three variables: age, fare, and cabin class (color can be either continuous or categorical here).
df.plot(kind = 'scatter', x='Fare', y='Age', c='Pclass', colormap='plasma', figsize=(15,4));
```

```python
# Using line plot for 2 cont var
# line chart: age by fare
df.groupby('Age')['Fare'].mean().plot(figsize=(12,4));
```

```python
# Line chart: age by fare and cabin class
mycolors=['crimson', 'purple', 'dodgerblue']
# Rather than specifying 3 plots (as we did above, with the histogram) you can just loop through the values.
for x in [1,2,3]:
    df[df['Pclass']==x].groupby('Age')['Fare'].mean().plot(figsize=(12,4), color=mycolors[x-1])
```

```python
# Comparing mult cont var
# Correlation Matrix
corr_matrix = df.corr()
corr_matrix
```

```python
# Display the correlation matrix as a heatmap
plt.imshow(corr_matrix);
```

```python
# More about seaborn further down.
sns.heatmap(corr_matrix);
```

```python
# Reproduce another example from FiveThirtyEight's shared data repository
# Chosen Data: Comic Books Are Still Made By Men, For Men And About Men
```

```python
# Reading the dataset
url = 'https://raw.githubusercontent.com/fivethirtyeight/data/master/comic-characters/dc-wikia-data.csv'
comic_dc = pd.read_csv(url)
comic_dc.head()
```

```python
# 'SEX' is the categorical variable and I shorted it to make the data easier to read
comic_dc['Sex_2'] = comic_dc['SEX'].map({'Male Characters':'Male', 'Female Characters':'Female'})
comic_dc.tail()
```

```python
# APPEARANCES is the continuous variable on this dataset, but contained strings, created another column to convert everything to int or NaN
comic_dc['Appearances1'] = pd.to_numeric(comic_dc['APPEARANCES'], errors='coerce')
```

```python
# APPEARANCES is a continuous variable, can be displayed as histogram
comic_dc['Appearances1'].plot(kind='hist');
```

```python
# removing outliers
comic_dc=comic_dc[comic_dc['Appearances1']<300]
comic_dc['Appearances1'].plot(kind='hist', color='royalblue');
```

```python
# converting continuous variables to categorical
appbins=[0,10,15,25,500]
applabels=['vcommon', 'common', 'uncommon', 'rare']
comic_dc['app_groups']= pd.cut(comic_dc['Appearances1'], bins=appbins, labels=applabels)
myresults = comic_dc['app_groups'].value_counts().sort_index()
myresults
```

```python
# display single continuous variable
comic_dc['Appearances1'].plot(kind='hist', bins=20, figsize=(16,4), color='yellow');
```

```python
# stacked histogram by sex and by appearance
comic_dc.head()
```

```python
comic_dc[comic_dc['Sex_2']=='Male']['Appearances1'].plot(kind='hist', bins=20, figsize=(16,4), alpha=0.5);
comic_dc[comic_dc['Sex_2']=='Female']['Appearances1'].plot(kind='hist', bins=20, figsize=(16,4), alpha=0.5);
```

```python
# using kernel density estimate
comic_dc[comic_dc['Sex_2']=='Male']['Appearances1'].plot(kind='density', xlim=(-5, 85), figsize=(16,4), color='navy')
comic_dc[comic_dc['Sex_2']=='Female']['Appearances1'].plot(kind='density', xlim=(-5, 85), figsize=(16,4), color='crimson');
```

```python
# Display a single categorical variable
```

```python
# Bar chart `pop` the rare category
comic_dc['app_groups'].value_counts().plot(kind='bar', color=['green', 'red', 'green', 'green']);
```

```python
# pie chart
comic_dc['app_groups'].value_counts().plot(kind='pie', figsize=(4,4));
```

```python
# bar chart of data above
char_id.plot(kind='bar');
```

```python
# crosstab with column percents
char_id1 = pd.crosstab(comic_dc['ID'],comic_dc['app_groups'], normalize='columns')
char_id1
```

```python
# stacked bar
char_id1.plot(kind='bar', stacked=True, color=['green', 'lime', 'lightgreen', 'teal']);
```

```python
# crosstab with column percents
char_id2 = pd.crosstab(comic_dc['ID'],comic_dc['app_groups'], normalize='index')
char_id2
```

```python
# paired barchart
char_id2.plot(kind='barh', color=['blue', 'royalblue', 'lightblue', 'navy'], figsize=(10,4));
```
