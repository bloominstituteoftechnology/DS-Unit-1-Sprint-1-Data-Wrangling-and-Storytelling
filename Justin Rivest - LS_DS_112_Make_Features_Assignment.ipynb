```python
# Create and output the dataframe.
df = pd.read_csv('LoanStats_2018Q4.csv', skiprows=1, skipfooter=2, engine="python")
df = df.drop(['id','member_id','desc','url'], axis=1)
df
```

```python
# Check datatypes.
print(df.shape)
df.dtypes
```

```python
# Test cleaning a single term.
df['term'][0].strip(" months")
```

```python
# Check for NaN values.
df['term'].isnull().sum()
```

```python
# Create a function that converts a term from a string to an integer and returns the int.
def termToInts(term):
  return int(term.strip(" months"))

# Test the function to make sure it works correctly.
termToInts("543 months")
```

```python
# Apply termToInts to every term in the 'term' column.
df['term'].apply(termToInts)
```

```python
# Overwrite the 'term' column with the new version.
df['term'] = df['term'].apply(termToInts)
```

```python
# Make sure it worked.
df['term'].dtypes
```

```python
# Outpute the first 5 instances
df.head()
```

```python
# Making the 'loan_status_is_great' column

# Check some details about the 'loan_status'
print(df['loan_status'].dtypes)
df['loan_status']
```

```python
# Create a function that returns 1 or 0 depending on the status of the loan.
def loanStatToInt(stat):
  if stat == "Fully Paid" or stat == "Current":
    return 1
  else:
    return 0

# Test the function
loanStatToInt("Fully Paid")
```

```python
df['loan_status_is_great'] = df['loan_status'].apply(loanStatToInt)
df
```

```python
# Make 'last_pymnt_d_month' and 'last_pymnt_d_year'  columns

# Checking details about the 'last_pymnt_d' column
df['last_pymnt_d']
```

```python
# Check for NaN's.
df['last_pymnt_d'].isnull().sum()
```

```python
# Drop NaN's
```

```python
# Convert the strings in the column to datetime objects.
df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], infer_datetime_format=True)
df['last_pymnt_d']
```

```python
# Make the 'last_pymnt_d_month' column.
df['last_pymnt_d_month'] = df['last_pymnt_d'].dt.month
```

```python
# Test the output and make sure it worked.
df['last_pymnt_d_month']
```

```python
# Make the 'last_pymnt_d_year' column.
df['last_pymnt_d_year'] = df['last_pymnt_d'].dt.year
```
