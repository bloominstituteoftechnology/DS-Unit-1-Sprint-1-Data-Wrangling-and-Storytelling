```python
# We want to drop the 100% blank columns

df = df.drop(['url', 'member_id', 'desc', 'id'], axis=1)
```

```python
cell = df['int_rate'][2]
type(cell)
# This column should not be a string.
```

```python
# We need to go about cleaning this column.

float(cell.strip().strip('%'))
```

```python
# Create a function for the .apply method

def clean_int(cell):
  return float(cell.strip().strip('%'))
```

```python
# ^Ok so we have managed to process one cell... now to apply to column.
# We can do this using the pd.apply() method.

df['int_rate_cleaned'] = df['int_rate'].apply(clean_int)

# Now we replace our working output column to the original int_rate column

df['int_rate'] = df['int_rate_cleaned']
```

```python
# Check type

type(df['int_rate'][2])
```

```python
# Drop working colum.

df.drop('int_rate_cleaned', axis=1, inplace=True)
```

```python
df['emp_title'].value_counts(dropna=False)[:20]
```

```python
# We want to look at how many unique employment titles exsist in this column.

print(f"Unique Values: {len(df['emp_title'].value_counts())}")
print(f"Total Observations: {len(df['emp_title'])}")
```

```python
# Number of null cases for emp_titile.
df['emp_title'].isnull().sum()
```

```python
# An easy way to approach this is to create a list of example'problem' inputs.

examples = ['owner', 'Supervisor', 'Project Manager', np.NaN]
```

```python
# Write a function to clean up these example cases.
# We are shooting for uniformity because this will allow us to more easily recognize (non-exact) duplicates.
# We return the input after passing it through the .title() method which creates uniformity in case level.
# We also strip any whitespace using the .strip() method

def clean_title(job_title):
  if isinstance(job_title, str):
    return job_title.title().strip()
  else:
    return "Unknown"
  
for title in examples:
  print(clean_title(title))


```

```python
# Here we do a list comprehension

[clean_title(example) for example in examples]
```

```python
# Now we want to apply this function to the emp_title column...
# We can do a direct replacement

df['emp_title'] = df['emp_title'].apply(clean_title)

df.head()
```

```python
# Let's see how much progress we made.

df['emp_title'].value_counts(dropna=False, ascending=False)[:20]
```
