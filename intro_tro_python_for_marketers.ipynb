### FILTERING AND RENAMING COLUMNS

```
# look at columns
# column name can be annoying, helps name them better
# also programing languages don't like white spaces in naming conventions
ksmdata.columns
```

```
# change the unnamed one into ID
ksmdata.rename(columns={'Unnamed: 0':'ID'}, inplace=True)
```

```
# inspect for change
ksmdata.head()
```

```
# how to replace annoying characters
# look up all the columns, replace anywhere you see '. repalce with ''(nothing)
# str is a string method
ksmdata.columns = ksmdata.columns.str.replace('.', '')

# inspect
ksmdata.head()
```

### FILTERING THE DATA SET 

```
# let's look at our features/columns
ksmdata.columns
```

```
# we want to filter our data to only show campaigns where
# amount pledged was more than/equals 100 dollars & goal was more than/equal to 100
# so basically people who met/exceeded their goals

# we're calling the data frame because we want to only show rows of the df
# that meets our specific criteria
ksmdata[(ksmdata['amtpledged'] >= 100) & (ksmdata['goal']>= 100)]
```

```
# store this filter
pledgegoal100 = ksmdata[(ksmdata['amtpledged'] >= 100) & (ksmdata['goal']>= 100)]

# check
pledgegoal100.head()
```

```
# looking for a combination of words in a series/column
# we want a data frame that only shows rows where the 'category' column contains
# 'fil' or the 'blurb' column contains 'love'
filmslove = ksmdata[(ksmdata['category'].str.contains('film') | ksmdata['blurb'].str.contains('love'))]
filmslove.head()
```

```
# filter data frame for rows where the category says 'video games' and the blurb
# says 'school'
ksmdata[ksmdata['category'].str.contains('Video Games') & ksmdata['blurb'].str.contains('school')].head()
```
