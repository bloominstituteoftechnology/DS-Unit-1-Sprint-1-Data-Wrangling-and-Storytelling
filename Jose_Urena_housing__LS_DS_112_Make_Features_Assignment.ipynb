# Assignment Part 1 workout

- Convert the `term` column from string to integer.


```python
df_housing['term'].value_counts()
```

```python
term_modify = '36 months'
```

```python
term_modify.strip('months').strip()
```

```python
type(term_modify)
```

```python
type(int(term_modify.strip('months').strip()))
```

```python
def term_modify_int(cell_contents):
  return int(cell_contents.strip('months').strip())
```

```python
df_housing['term'] = df_housing['term'].apply(term_modify_int)
```

```python
df_housing['id'].dtypes
```

```python
#Make a column named loan_status_is_great. It should contain the integer 1 if loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.
def great(loan_status):
    if loan_status == 'Current' or loan_status == 'Fully Paid':
        return 1
    else:
        return 0
    
df_housing['loan_status_is_great'] = df_housing['loan_status'].apply(great)
```

```python
print(df_housing['loan_status_is_great'].value_counts())
print(df_housing['loan_status'].value_counts())
```
