```python
# Your Work Here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('fivethirtyeight')
```

```python
from IPython.display import display, Image
url = 'https://fivethirtyeight.com/wp-content/uploads/2018/10/actblue-1.png?w=575'
example = Image(url=url, width=400)
display(example)
```

```python
fake2 = pd.Series([4,6,15],
                  index=range(1,4))
fake2.plot.bar();
```

```python
plt.style.use('fivethirtyeight')

fig, ax = plt.subplots()

#fig style to white
fig.set_facecolor('white')
#axes styles
ax.bar(x=[1,2,3], height=fake2, width=.9, color='lightgreen')
ax.set_facecolor('white')

#title and subtitle
ax.set_title(x=.9, y=1.5, label='ActBlue has only gotten more popular', fontsize=14, fontweight='bold')
ax.text(x=.9, y=21, s='Number and total dollar amount of contributions made via ActBlue to', fontsize=12, color='black');
ax.text(x=1.5, y=20, s='House and Senate candidates in two-year period', fontsize=12, color='black');
ax.text(x=.89, y=17, s='Number of contributions', fontsize=12, color='black', fontweight='bold');
ax.text(x=0, y=-5, s="'For 2014 and 2016 data covers Jan. 1 of the year before the election through Dec. 31 of the election", fontsize=12, color='grey');
ax.text(x=0, y=-6, s='year. For 2018, data covers Jan. 1, 2017, throught Seept. 30, 2018', fontsize=12, color='grey');
ax.text(x=0, y=-8, s='FiveThirtyEight', fontsize=12, color='grey');
ax.text(x=3.3, y=-8, s='SOURCE FEDERAL EDUCATION COMMISSION', fontsize=12, color='grey');
#tick marks 
ax.set_xticks([1,2,3]);
ax.set_yticks([0,5,10,15]);
#tick labels
ax.set_xticklabels([2014,"'16","'18"])
ax.set_yticklabels([0,5,10,'15m'], fontsize=13, color='grey')
#spine
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_facecolor('black') 
#grid color
ax.grid(color='lightwhite', axis='both', which='minor')
#hide x axis
plt.grid(b=None, axis='x')





plt.show()
```
