Let's convert the 'term' column wich type is a string into an integer. I am going to build a function that will get rid of the space and the % symbol (for each cel). And then, apply it to the whole column.

```python
def term_into_integer (string):
    result =''
    for a in string:
        if a in [' ','m', 'o', 'n', 't', 'h', 's']:
           result = result
        else:
           result += a
    return int(result)
```

Let see if it works on an example.

It does, let's run it now over the whole 'term' column

```python
grocery['terms_in_months'] = [term_into_integer(code) for code in grocery['term']]
grocery['terms_in_months'].head()
```

I want to build a function that reads a string and returns 1 if the read string is wether 'Current' or 'Fully paid' or 0 otherwise.

```python
def is_current_or_fully_paid (string):
    if string == 'Current' or string == 'Fully Paid':
       return 1
    else:
       return 0
```

Let's now create a new column 'loan_status_is_great', with 1 in it for loan_status being 'Current' or 'Fully Paid' and 0 otherwise.

```python
grocery['loan_status_is_great'] = grocery['loan_status'].apply(is_current_or_fully_paid)
grocery.head()
```

Let's change the type of 'next_pymnt_d' and 'last_pymnt_d' columns into datetime type.

```python
grocery['next_pymnt_d'] = pd.to_datetime(grocery['next_pymnt_d'], infer_datetime_format=True)
grocery['next_pymnt_d'].head()
```

```python
grocery['last_pymnt_d'] = pd.to_datetime(grocery['last_pymnt_d'], infer_datetime_format=True)
grocery['last_pymnt_d'].head()
```
