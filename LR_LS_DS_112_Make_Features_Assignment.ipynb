```python
# Create some examples that represent the cases that we want to clean up
examples = ['owner', 'Supervisor', ' Project Manager', np.nan]
```

```python
# Write a function to clean up these use cases and increase uniformity.
# isinstance is a method that python uses to determine if a given value belongs to a certain datatype 
def clean_title(job_title):
  if isinstance(job_title, str):
    return job_title.title().strip()

  else:
    return "Unknown"
```

```python
#create a for loop
for title in examples:
  print(clean_title(title))
```

```python
# list comprehensions can combine function calls and for loops over lists
# into one succinct and fairly readable single line of code.
[clean_title(title) for title in examples]
```

```python
# We have a function that works as expected. Lets apply it to our column.
# This time we'll overwrite the original column
df['emp_title'] = df['emp_title'].apply(clean_title)
df.head()
```

```python
#write a function to check for current and fully paid
numeric_status = []
def check_loan_status(cells):
  if cells == "Current":
    return 1
  elif cells == "Fully Paid":
    return 1
  else:
    return 0
```

```python

for cell in df['loan_status']:
  numeric_status.append(check_loan_status(cell))
print(numeric_status)
```

```python
#using apply function to create new column
df['loan_status_is_great'] = df['loan_status'].apply(check_loan_status)
df['loan_status_is_great'].head(20)
```
