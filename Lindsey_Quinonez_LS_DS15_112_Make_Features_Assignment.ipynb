```python
##CREATING FUNCTION TO STRIP int_rate VALUES OF THEIR WHITESPACE AND '?'##
def int_rate_to_float(dirty_int_rate):
  return float(dirty_int_rate.strip().strip('%'))
```

```python
## CASTING int_rate COLUMN VALUES TO A FLOAT##
Loan_Stats['int_rate'].apply(int_rate_to_float)                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                            
```

```python
##HERE, I AM CREATING A NEW COLUMN TITLED 'int_rate_cleaned' AND ASSIGNING IT 
#THE "NEW" FLOAT VALUES 
Loan_Stats['int_rate_cleaned'] = Loan_Stats['int_rate'].apply(int_rate_to_float) 
Loan_Stats.head()
```

```python
#Create some examples that represent the cases that we want to clean up
import numpy as np

examples = ['owner', 'Supervisor', ' Project Manager', np.NaN]
```

```python
 #Write a function to clean up these cases and increase uniformity.
 #The error shows me that np.Nan is NOT a string and therefore cannot be manipulated with .strip()
 #A np.NaN value is a float
 #Type checking needs to be completed. isinstance()
 # .title() - Title Casing allows us to uppercase the first letter of every word. and makes sure there is consistent casing throughout.
def clean_emp_title(title):
  if isinstance(title, str):
    return title.strip().title()
  else:
    return "Unknown"  
for example in examples:
  print(clean_emp_title(example))
```
