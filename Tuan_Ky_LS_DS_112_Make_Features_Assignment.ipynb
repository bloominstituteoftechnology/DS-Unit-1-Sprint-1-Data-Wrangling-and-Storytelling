```python
import pandas as pd 
pd.options.display.max_rows = 999
pd.options.display.max_columns = 100

df = pd.read_csv('LoanStats_2018Q4.csv', skiprows = 1, skipfooter = 2, engine = 'python')
print(df.shape)
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum().sort_values()
```

```python
# drop columns that are made up completely of NaN values

df = df.drop(['url', 'member_id', 'desc', 'id'], axis=1)
```

```python
df.head()
```

```python
term = ' 36 months'
term
```

```python
term.strip()
```

```python
term.strip().strip(' months')
```

```python
int(term.strip().strip(' months'))
```

```python
def term_to_int(mychar):
  return int(mychar.strip().strip(' months'))

term_to_int(term)
```

```python
#apply function to term column

df['term'].apply(term_to_int)[:10]
```

```python
df['term'] = df['term'].apply(term_to_int)
df.head()
```

```python
#Make a column named loan_status_is_great. It should contain the integer 1 if loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.
loan_status = "Fully Paid"
loan_status1 = "Current"
loan_status2 = "IDK"
def loan_status_is_great(forsure):
  if forsure == "Current":
    forsure = 1
  elif forsure == "Fully Paid":
    forsure = 1
  else:
    forsure = 0
  return forsure
loan_status_is_great(loan_status1)
```

```python
df['loan_status_is_great'] = df['loan_status'].apply(loan_status_is_great)
df.head()
```

```python
df['loan_status'].value_counts()
#checking to see if there are other query other than current or fully paid since head(10) showing all 1
```

```python
#Make last_pymnt_d_month and last_pymnt_d_year columns.
df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], infer_datetime_format = True)
df['last_pymnt_d_month'] = df['last_pymnt_d'].dt.month
df['last_pymnt_d_year'] = df['last_pymnt_d'].dt.year
```
