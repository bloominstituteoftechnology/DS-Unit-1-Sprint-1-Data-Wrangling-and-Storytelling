```python
# In the int_rate column, we have a few problems.
# The values are considered strings, the % is included, and there is a leading space
# To remove these obstacles we will have to take care of them 1 at a time
# .strip() remvoves certain things from a string
# .strip() removes leading spaces & .strip(%) removes the % sign
int_rate = ' 14.17%'
int_rate.strip().strip('%')
```

```python
# We can also create a function that can do the same thing with any string
def int_rate_to_float(cell_contents):
  return float(cell_contents.strip().strip('%'))
```

```python
# We can take a look at some examples of Employer Titles that we can change
# We can also clean them up so that they have the same formatting
import numpy as np

examples = ['owner', 'Supervisor', ' Project Manager', np.NaN]

def clean_title(job_title):
  if isinstance(job_title, str):
    return job_title.title().strip()
  else:
    return "Unknown"
  
for title in examples:
  print(clean_title(title)) #This reformats things using "Title casing", like a book's title
```

```python
# A List Comprehenstion is a much more succinct and readable way of writing code
# They combine function calls and for loops over lists

[clean_title(example) for example in examples]
#In plain English, this can be read as "Clean the Title for each example in examples"
```

```python
# Here we start the new function string to integer
# For this it is mean to remove the white space and the word months
# in a cell
def string_to_int(cell_contents):
  return int(cell_contents.strip().strip('months'))
```

```python
# This states that for every cell in the Term column, apply the aforementioned
# function and print
# The result is the strings having been casted into integers
for cell in df['term']:
  print(string_to_int(cell))
```

```python
# Next we will try to Make a column named loan_status_is_great. It should 
# contain the integer 1 if loan_status is "Current" or "Fully Paid." 
# Else it should contain the integer 0.
df['loan_status'].value_counts(dropna = False)[:20]
```

```python
# Here we are starting a new function to turn the loan status names
# into an integer, either 1 or 0
def loan_status_to_integer(loan_status):

# This is stating that if the loan status reads as Current or Fully Paid
# Then return the integer 1
  if (loan_status == 'Current') or (loan_status == 'Fully Paid'):
    return 1

# Otherwise, return the integer 0
  else:
      return 0

# This is creating the new column "loan_status_is great", stating
# that it is equivalent to the "loan_status column" applying the loan_status_to_integer function

df['loan_status_is_great'] = df['loan_status'].apply(loan_status_to_integer)
df.head(10)
```
