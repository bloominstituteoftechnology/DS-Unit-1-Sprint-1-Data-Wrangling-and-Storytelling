## 2.1 What are the dimensions of the following matrices? Which of the following can be multiplied together? See if you can find all of the different legal combinations.
\begin{align}
A = \begin{bmatrix}

\qquad
D = \begin{bmatrix}

0 & 0 & 1
\end{bmatrix}
\qquad
E = \begin{bmatrix}
1 & 3 \\
5 & 7
\end{bmatrix}
\end{align}

```
# Matrices
A = np.array([[1, 2], [3, 4], [5, 6]])
B = np.array([[2, 4, 6]])
C = np.array([[9, 6, 30], [4, 7, 11]])
D = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
E = np.array([[1, 3], [5, 7]])

# Dimensions of Matrices
print('Matrix A Dimensions:', np.shape(A))
print('Matrix B Dimensions:', np.shape(B))
print('Matrix C Dimensions:', np.shape(C))
print('Matrix D Dimensions:', np.shape(D))
print('Matrix E Dimensions:', np.shape(E))
```

```
# Legal Combinations
m1 = np.matmul(A, C)
m2 = np.matmul(A, E)
m3 = np.matmul(B, A)
m4 = np.matmul(B, D)
m5 = np.matmul(C, A)
m6 = np.matmul(C, D)
m7 = np.matmul(D, A)
m8 = np.matmul(E, C)

# Results
print('Matrix A * Matrix C:', m1)
print('Matrix A * Matrix E:', m2)
print('Matrix B * Matrix A:', m3)
print('Matrix B * Matrix D:', m4)
print('Matrix C * Matrix A:', m5)
print('Matrix C * Matrix D:', m6)
print('Matrix D * Matrix A:', m7)
print('Matrix E * Matrix C:', m8)
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
# CD
m6 = np.matmul(C, D)
print('Dimension:', m6.shape)
print(m6)

# AE
m2 = np.matmul(A, E)
print('Dimension:', m2.shape)
print(m2)

# BA
m3 = np.matmul(B, A)
print('Dimension:', m3.shape)
print(m3)
```

The resulting matrices are equal to the first matrix's row total and the second matrix's column total.

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = np.array([[12, 11], 
              [7, 10]])

J = np.array([[0, 1, 2],
              [7, 10, 4], 
              [3, 2, 0]])

print('Determinant of H:', np.linalg.det(H))
print('Determinant of J:', np.linalg.det(J))
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
print('Inverse of H:', np.linalg.inv(H))
print('Inverse of J:', np.linalg.inv(J))
```

## 3.4 Find $HH^{-1}$ and then find $J^{-1}J$. Is $HH^{-1} == J^{-1}J$? Why or Why not? 

Please ignore Python rounding errors. If necessary, format your output so that it rounds to 5 significant digits (the fifth decimal place).

```
H_x = np.linalg.inv(H)
HHinv = np.matmul(H, H_x)

J_x = np.linalg.inv(J)
JinvJ = np.matmul(J_x, J)

print(HHinv)
print(JinvJ)
```

Is $HH^{-1} == J^{-1}J$? YES, because any number multiplied by its inverse = 1, and 1 == 1.
