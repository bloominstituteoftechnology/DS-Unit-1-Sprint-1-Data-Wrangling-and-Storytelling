```python
#ASSIGNMENT: Make a column named loan_status_is_great. 
#It should contain the integer 1 if loan_status is "Current" or "Fully Paid." 
#Else it should contain the integer 0.

#Here is the test
loan_test = ['Current']


def binary_loan_function(status):
  if status in ['Current','Fully Paid']:
    return 1
  else:
    return 0

for status in loan_test:
  print(binary_loan_function(status))
```

```python
#Here is the creation of the new column

df['loan_status_is_great'] = df['loan_status'].apply(binary_loan_function)

df.head()
df['loan_status_is_great'].value_counts()
df['loan_status_is_great'][-20:]
```

```python
#Make last_pymnt_d_month and last_pymnt_d_year columns.

df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], infer_datetime_format=True)
df['last_pymnt_d'].head()

df['lastpymnt_d_month'] = df['last_pymnt_d'].dt.month.astype('Int64')
df['lastpymnt_d_month'].head()

df['lastpymnt_d_year'] = df['last_pymnt_d'].dt.year.astype('Int64')

df['lastpymnt_d_year'].head()
```
