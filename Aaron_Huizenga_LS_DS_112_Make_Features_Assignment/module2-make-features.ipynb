```python
# The object now is to create a column "loan status is great" with it containing
# the int 1 "if" loan status is "current" or "fully paid" Else containing int 0

df['loan_status_is_great'] = df['loan_status']
df.columns
```

```python
# We need to create function for new column

def loan_status_is_great(loan_status):
  if (loan_status =='Current') or (loan_status =='Fully Paid'):
    return '1'
  else:
    return '0'
print(loan_status_is_great)
```
