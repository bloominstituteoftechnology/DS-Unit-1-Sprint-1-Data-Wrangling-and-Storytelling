```python
import pandas as pd
import numpy as np

table1 = pd.DataFrame(
    [[np.nan, 2],
     [16,    11], 
     [3,      1]],
    index=['John Smith', 'Jane Doe', 'Mary Johnson'], 
    columns=['treatmenta', 'treatmentb'])

table2 = table1.T

table2
```

# table2 -> tidy

```python
#Take the row index, and add it as a new column
table2 = table2.reset_index()
table2
```

```python
# What the unique identifier for each row
# Where is the data at that I want to be in my single "tidy" column
# MELT FUNCTION - Go from WIDE -> TIDY

tidy2 = table2.melt(id_vars='index', value_vars=['John Smith', 'Jane Doe', 'Mary Johnson'])
tidy2
```

```python
tidy2 = tidy2.rename(columns={
    'index': 'trt',
    'variable': 'name',
    'value': 'result'
})
tidy2
```

```python
tidy2.trt = tidy2.trt.str.replace('treatment', '')

tidy2
```

```python
tidy2.set_index([pd.Index([1, 2, 3, 4,5,6]), 'name'])
```

# Tidy -> Table2

```python
#Using the pivot_table function (It's inverse of melt.)

# index: unique identifier
# columns: What di you want to differentiate the columns in wide format;
# values: Where are the numbers at - go in the middle of the wide dataframe.

wide = tidy2.pivot_table(index='trt', columns='name', values='result')

wide
```
