## Practice 1-Sample T-tests

One Sample t-tests determine whether or not a sample mean is statistically different from some known (or hypothesized) population mean. 

### 1) Load the Data
- Use the [automobile dataset](https://archive.ics.uci.edu/ml/datasets/Automobile)
- Fix the column headers
- Make sure NaNs are used to indicate missing values

Feel free to add code cells and text cells as needed throughout the assignment.

```python
import pandas as pd
import numpy as np

url = 'https://archive.ics.uci.edu/ml/datasets/Automobile'
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'
df = pd.read_csv(url)
column_names = ['symboling', 'normalized-losses', 'make', 'fuel-type', 
                'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 
                'engine-location', 'wheel-base', 'length', 'width', 'height', 
                'curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size', 
                'fuel-system', 'bore', 'stroke', 'compression-ratio', 
                'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
df.columns = column_names
df.head(20)
```

```python
df = df.replace('?', np.nan)
df.head(20)
```

```python
df.isnull().sum()
```

### 2) Pretend that this dataset represents the cars at a used car lot in your local town. 

- Use df.sample() to pick a random sample of 10 cars. Note that because this sample is random we are going to set the `random_state` so that all of us in the class get the same random sample. Please set your random state to `30` when using `df.sample()`

Is your sample reflective of the population value in regards to highway-mpg? Find the mean for `highway-mpg` for the entire dataset and use a 1-sample t-test to compare your estaimated sample mean to the population mean. Can you say that they are different? 

```python
from scipy import stats
```

```python
used_lot = df.sample(n= 10, random_state= 30)
# mean value for all of highway-mpg values
print('population ', df['highway-mpg'].mean())
# mean value for used car lot mpg
stats.ttest_1samp(a= used_lot['highway-mpg'], popmean= 30.769)
```

###The sample and the population are different since the p-valus is low at 0.74

The salesman says the cars he sells typically have a fuel efficiency of about 35 miles per gallon on the highway. You want to verify his claim but can only test 10 cars. Using your sample of 10, test his claim and report your results.

```python
print(stats.ttest_1samp(used_lot['highway-mpg'], 35))
used_lot['highway-mpg'].mean()
```

### T-test shows that the salesman's claim is false because of a very low p-value

Imagine that you now have the capacity to test 100 cars on the lot. Using the same random state of `30`, take a sample of 100 cars. Run a t-test to verify the salesman's claim with your new larger sample. Do you reach the same conclusion as you did with the sample size of 10? 

```python
used_lot_100_cars = df.sample(n= 100, random_state= 30)
```

```python
stats.ttest_1samp(a= used_lot_100_cars['highway-mpg'], popmean= 35)
```

Why might these two t-tests using the same dataset lead to different conclusions about the salesman's claim?

###P-value for the sample of one hundred cars is much smaller than the sample of ten cars which shows the saleman's claim is even more false. T-statistic has also increased from -3.418 to -6.482 which shows that the null hypothesis of 35 is even more incorrect.
