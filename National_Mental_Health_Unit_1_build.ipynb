```
mh_youth_suicide_f = mh_df_youth[mh_df_youth["MHSUICIDE"]==1]
#filter df for features of age < 17 years & suicide prevention services
mh_youth_suicide_f
mh_youth_suicide_n = len(mh_youth_suicide_f["MHSUICIDE"]) #filter for facilities with youth facilities with suicide prevention services.
```
