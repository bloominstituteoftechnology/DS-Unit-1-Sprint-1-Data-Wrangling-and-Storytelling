## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors?

## 1.5 find $\vec{a} - \vec{b}$ and plot the result on the same graph as $\vec{a}$ and $\vec{b}$. Is there a relationship between vectors $\vec{a} \thinspace, \vec{b} \thinspace \text{and} \thinspace \vec{a-b}$

```
a = [5,7]
b = [3,4]

# Turn vector lists into numpy arrays
vec_a = np.array(a)
vec_b = np.array(b)

minus = vec_a - vec_b

plt.arrow(0,0, a[0], a[1], head_width=.30, head_length=.45, 
          color='red')
plt.arrow(0,0, b[0], b[1], head_width=.30, head_length=.45, 
          color='green')

# Plot the a-b
plt.arrow(0,0, minus[0], minus[1], head_width=.30, head_length=.45, 
          color='purple')

plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

```
'''
Vector a-b appears to go in the same direction but is smaller, also looks to 
have a higher "trajectory" (not sure if that is an appropriate way to describe,
but essentially I'm saying this arrow seems to point more upward)
'''
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
A = np.array([[1,2],
              [3,4],
              [5,6]])

B = np.array([2,4,6])

C = np.array([[9,6,3],
             [4,7,11]])

D = np.array([[1,0,0],
              [0,1,0],
              [0,0,1]])

E = np.array([[1,3],
              [5,7]])

# CD
print('CD:',np.matmul(C,D))
#AE
print('AE:',np.matmul(A,E))
#BA
print('BA:',np.matmul(B,A))

'''
CD is a 2x3 matrix, AE is 3x2 and BA is 1x1. When put side by side, the
dimension of the product matrix is the number of rows from the first factor
matrix and the number of columns from the second factor matrix. For example, 
in CD, we multiplied a 2x3 matrix with a 3x3 matrix and got a 2x3 result.
'''
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = np.array([[12,11],
              [7,10]])

J = np.array([[0,1,2],
              [7,10,4],
              [3,2,0]])

print('Determinant of H:',np.linalg.det(H))
print('Determinant of J:',np.linalg.det(J))
```
