## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
# Create matrices
A = np.array([[1, 2],
             [3, 4],
             [5, 6]])

B = np.array([2, 4, 6])

C = np.array([[9, 6, 3],
             [4, 7, 1]])

D = np.array([[1, 0, 0],
             [0, 1, 0],
             [0, 0, 1]])

E = np.array([[1, 3],
             [5, 7]])

# CD
print(np.matmul(C, D)) # 2 x 3
print(np.matmul(A, E)) # 3 x 2
print(np.matmul(B, A)) # 1 x 2
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
# Create matrices

H = np.array([[12, 11],
             [7, 10]])

J = np.array([[0, 1, 2],
             [7, 10, 4],
             [3, 2, 0]])

# Find determinants
print(np.linalg.det(H))
print(np.linalg.det(J))
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
# Find inverse
H_inv = np.linalg.inv(H)
J_inv = np.linalg.inv(J)

print(H_inv)
print(J_inv)
```
