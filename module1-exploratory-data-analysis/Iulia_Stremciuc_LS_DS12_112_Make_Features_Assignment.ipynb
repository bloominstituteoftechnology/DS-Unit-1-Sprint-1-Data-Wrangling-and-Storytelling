```python
term = ' 60 months'
```

```python
def term_to_int(cell_contents):
  return int(cell_contents.strip(' months'))

term_to_int(term)
```

```python
my_list = []
for cell in df['term']:
  my_list.append(term_to_int(cell))

my_list[:15]
```

```python
df['term_cleaned'] = pd.Series(my_list)
df.head()
```

```python
df['term_cleaned_apply_form'] = df['term'].apply(term_to_int)
df.head()
```

```python
df['loan_status'].head()
```

```python
##Make a column named loan_status_is_great. It should contain the integer 1 if loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.

def great_status(status):
  if status == 'Current' or status =='Fully Paid':
    return 1
  else:
    return 0

for status in df['loan_status']:
  print(great_status(status))  
  
```

```python
df['loan_status_is_great'] = df['loan_status'].apply(great_status)
df.head()
```
