```python
examples = ['owner', 'Supervisor', ' Project Manager', np.NaN]

#write a cleaning and uniformity function
def clean_title(job_title):
  if isinstance(job_title, str):
    return job_title.title().strip()
  else:
    return "Unknown"

for title in examples:
  print(clean_title(title))
```

```python
#make a list that does the same thing as above using one line of code
[clean_title(title) for title in examples]
```

```python
#apply this function to the whole column
df['emp_title']=df['emp_title'].apply(clean_title)

df.head()
```

```python
##Convert the 'term' column from string to integer.

#taking a look at the term column
print(df['term'].head(10))

#taking a look at a specific entry to see what needs to be stripped to convert the string to an integer
print(df['term'][0])

#figuring out my line of code on a test subject
test_term = df['term'][0]
print(test_term)
test_term.strip('months').strip()
int(test_term.strip('months').strip())
type(int(test_term.strip('months').strip()))

#writing a function to apply this to all the data in the column

def string_to_integer(cellcontents):
  return int(cellcontents.strip('months').strip())

df['int_term']=df['term'].apply(string_to_integer)

df['int_term'].head(10)
```

```python
#Make a column named loan_status_is_great. It should contain the integer 1 if loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.



#checking out the column
df['loan_status'].head(10)

def loan_status_helper(cell):
  if cell == "Current" or cell == "Fully Paid":
    return 1
  else: 
    return 0

df['loan_status_is_great'] = df['loan_status'].apply(loan_status_helper)


df['loan_status_is_great'].head(10)


#did I accidentally change the 'loan_status' column as I was doing this?

df['loan_status'].head(100)



```
