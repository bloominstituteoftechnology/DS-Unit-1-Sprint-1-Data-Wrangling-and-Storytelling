# Make a column named loan_status_is_great. It should contain the integer 1 if loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.

```python
def conditions(f):
    if (f['loan_status']  == "Current") or (f['loan_status'] == 'Fully Paid'):
        return 1
    else:
        return 0


df['loan_status_is_great'] = df.apply(conditions, axis=1)
df.head()
```
