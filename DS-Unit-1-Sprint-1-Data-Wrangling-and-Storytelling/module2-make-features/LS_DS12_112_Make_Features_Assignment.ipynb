```python
#applying this solution to every cell by writing
#a reusable function, but in this case, for the integer conversion of the term
#column, per the assignment

#cell_contents is a variable or parameter for/or casted in the variable term_to_int?
#the 'return' directive casts the cell_contents variable with chained methods
# and parameters into INT.


def term_to_int(cell_contents):
  return int(cell_contents.strip().strip('months'))

```

```python
#testing this to see if it works on the single variable from earlier
term_to_int(term_conversion)
```

```python
#applying this function whose variable is called, term_to_int, to every cell in the !!!!!!term column!!!!!!

new_list= []

for cell in df['term']:
  new_list.append(term_to_int(cell))

#Looking at first 10 values of the list
new_list[0:10]
```

```python
# Taking my list and turning it into a "Series" object 
# (the technical name for a dataframe column)
# And then I can save my new "Series" (column) as a new column.

#new column of 'term' at the end of the dataframe

df['int_term_cleaned'] = pd.Series(new_list)
df.head()
```

```python
#what is the datatype in this new column? It's an int. So we're good. 

df.dtypes[-5:]
```

```python
#improving the code with the apply() function while adding another column with same info.

df['int_term_cleaned_from_apply'] = df['term'].apply(term_to_int)

df.head()
```

```python
import numpy as np
#why might some employ titles not provided?
examples = ['owner', 'Supervisor', ' Project Manager', np.NaN]

#function to clean these up

#

def clean_title(job_title):
  if isinstance(job_title, str):
    return job_title.title().strip()
  else:
    return "Unknown"

#loop to run over examples list
for title in examples:
  print(clean_title(title))
```

```python
#list comprehensions can combine function calls and for loops over lists
#into one succinct and fairly readable single line of code.

#for loop on right, clean_title function on left

[clean_title(title) for title in examples]
```

```python
#We have a function that works as expected. Lets apply it to our column.
# This time we'll overwrite the original column

df['emp_title'] = df['emp_title'].apply(clean_title)

df.head()
```

Because all of these dates come from Q4 of 2018, the `issue_d` column isn't all that interesting. Lets look at the `earliest_cr_line` column, which is also a string, but that could be converted to datetime format.

We're going to create a new column called `days_from_earliest_credit_to_issue`

It's a long column header, but think about how valuable this piece of information could be. This number will essentially indicate the length of a person's credit history and if that is correlated with repayment or other factors could be a valuable predictor!

```python
df['earliest_cr_line'].head()
```

```python
df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'], infer_datetime_format=True)  
```

What we're about to do is so cool! Pandas' datetime format is so smart that we can simply use the subtraction operator `-` in order to calculate the amount of time between two dates. 

Think about everything that's going on under the hood in order to give us such straightforward syntax! Handling months of different lengths, leap years, etc. Pandas datetime objects are seriously powerful!

```python
df['issue_d'] - df['earliest_cr_line']
```

```python
df['days_from_earliest_credit_to_issue'] = (df['issue_d'] - df['earliest_cr_line']).dt.days

df.head()
```
