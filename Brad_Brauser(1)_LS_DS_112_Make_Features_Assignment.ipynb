```python
great_loan_statuses = ['Current', 'Fully Paid']

titles = ['Fully Paid',
          'Current', 
          'Charged Off', 
          'In Grace Period', 
          'Late (31-120 days)', 
          'Default', 
          'Late (16-20 days)']

def loan_status_is_great(title):
  if title not in great_loan_statuses:
    return 0
  return 1
```
