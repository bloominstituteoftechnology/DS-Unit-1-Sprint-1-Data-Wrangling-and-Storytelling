```python
import pandas as pd
import numpy as np

table1 = pd.DataFrame(
    [[np.nan, 2],
     [16,    11], 
     [3,      1]],
    index=['John Smith', 'Jane Doe', 'Mary Johnson'], 
    columns=['treatmenta', 'treatmentb'])

table2 = table1.T

table2
```

```python
table2 = table2.reset_index()
table2
```

```python
# table2 -> tidy

tidy2 = table2.melt(id_vars='index', value_vars=['John Smith', 'Jane Doe',
                                                 'Mary Johnson'])

tidy2
```

```python
tidy2 = tidy2.rename(columns={
    'index': 'trt', 
    'variable': 'name', 
    'value': 'result'
})
tidy2
```

```python
tidy2.trt = tidy2.trt.str.replace('treatment', '')

tidy2
```

```python
# tidy -> table2

wide2 = tidy2.pivot_table(index='trt', columns='name', values='result')
wide2
```
