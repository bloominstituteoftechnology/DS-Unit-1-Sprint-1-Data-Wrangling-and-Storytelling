## Top 10 Most Frequently Ordered Products
We need product names and counts of how many times they were sold
- product_name is in products.csv
- instances of specific orders are in order_products__prior.csv and order_products__train.csv

```python
import pandas as pd

orders = pd.read_csv('orders.csv')
orders.head()
```

We actually don't need anything from orders.csv

```python
products = pd.read_csv('products.csv')
products.head()
```

```python
# read in and concatenate the __prior and __train CSVs
order_products__prior = pd.read_csv('order_products__prior.csv')
order_products__train = pd.read_csv('order_products__train.csv')

order_products = pd.concat([order_products__prior, order_products__train])
print(order_products.shape)
order_products.head()
```

```python
order_products['product_id'].value_counts()
```

### Try and eyeball duplicate products in a single order

```python
pd.set_option('display.max_rows', 500)
order_products.sort_values(['order_id', 'product_id', 'reordered']).head(100)
```

### I'm not seeing any duplicates, but maybe we're just getting unlucky? Can we check for duplicates programmatically?

```python
order_products.duplicated(subset=['order_id', 'product_id']).value_counts()
```

```python
# order_products[['order_id', 'product_id']].nunique()
```

```python
order_products.shape
```

```python
order_products.duplicated(subset=['order_id', 'product_id']).value_counts().iloc[0]
```

```python
assert order_products.duplicated(subset=['order_id', 'product_id']).value_counts().iloc[0] == order_products.shape[0]
```

### Conclusion? - This dataset does not have any information about the quantity of items ordered, only unique items ordered and whether the shopper had bought any items in past visits. So our counts of how many times the top 10 products were ordered will really be the number of orders that the top 10 products were included in.

### In order to count the frequency of orders of a given product we need to combine orders and products so that we have names associated with the products in each order.

```python
product_orders = pd.merge(order_products, products[['product_id', 'product_name']])
product_orders.head()
```

### Getting the final counts is all about just counting up the number of unique instances of product_name

```python
product_orders['product_name'].value_counts().sort_values(ascending=False)[:10]
```

```python
product_orders['product_name'].value_counts().nlargest(10)
```

```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns

table1 = pd.DataFrame(
    [[np.nan, 2],
     [16,    11], 
     [3,      1]],
    index=['John Smith', 'Jane Doe', 'Mary Johnson'], 
    columns=['treatmenta', 'treatmentb'])

table2 = table1.T

table2
```

## Table2 --> Tidy

### Option 1, just transpose it (to get back to the table1 version) and then do the same thing you did to table1

```python
tidy2 = table2.T.reset_index().melt(id_vars='index').rename(columns= {
    'index': 'name', 
    'variable': 'trt', 
    'value': 'result'
})
tidy2
```

```python
tidy2['trt'] = tidy2['trt'].str.replace('treatment', '')
tidy2 = tidy2.set_index('name')
tidy2
```

## Tidy --> Table2

Do the same thing you did to table 1, but then transpose it when you're finished

```python
tidy2.pivot_table(index='name', columns='trt', values='result').T
```

### I given an example of both [pd.pivot](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.pivot.html) and [pd.pivot_table](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.pivot_table.html) in this instance the both give the same result but that is not always the case.

### `pivot_table` is the more powerful version of the function as it can handle multiple indexes (called a hierarchical index) and duplicate-value row/column combinations.
