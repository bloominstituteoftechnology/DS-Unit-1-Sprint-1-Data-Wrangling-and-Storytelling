## Practice 1-Sample T-tests

One Sample t-tests determine whether or not a sample mean is statistically different from some known (or hypothesized) population mean. 

### 1) Load the Data
- Use the [automobile dataset](https://archive.ics.uci.edu/ml/datasets/Automobile)
- Fix the column headers
- Make sure NaNs are used to indicate missing values

Feel free to add code cells and text cells as needed throughout the assignment.

```python
### YOUR WORK HERE
import pandas as pd
import numpy as np
from scipy import stats

column_names = ['symboling', 'normalized-losses', 'make', 'fuel-type', 
                'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 
                'engine-location', 'wheel-base', 'length', 'width', 'height', 
                'curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size', 
                'fuel-system', 'bore', 'stroke', 'compression-ratio', 
                'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data',
                 names=column_names)
df
```

```python
df = df.replace('?',np.NaN)
df
```

### 2) Pretend that this dataset represents the cars at a used car lot in your local town. 

- Use df.sample() to pick a random sample of 10 cars. 

---

Note that because this sample is random we are going to set the `random_state` so that all of us in the class get the same random sample. Please set your random state to `30` when using `df.sample()`

Is your sample reflective of the population value in regards to highway-mpg? Find the mean for `highway-mpg` for the entire dataset and use a 1-sample t-test to compare your estaimated sample mean to the population mean. Can you say that they are different? 

```python
### YOUR WORK HERE
sample = df.sample(10,random_state=30)
sample
```

The salesman says the cars he sells typically have a fuel efficiency of about 35 miles per gallon on the highway. You want to verify his claim but can only test 10 cars. Using your sample of 10, test his claim and report your results.

###Null Hypothesis:

The cars the salesman sells typically have a fuel efficiency of about 35 miles per gallon on the highway.

###Alternative Hypothesis:

The cars the salesman sells **do not** typically have a fuel efficiency of about 35 miles per gallon on the highway.

### Confidence Level: 95%

```python
### YOUR WORK HERE
stats.ttest_1samp(sample['highway-mpg'],35,nan_policy='omit')
```

###Conclusion:
Based on a t-statistic of -0.96, and a p-value of 0.36, I (reject/**fail to reject**) the null hypothesis that the cars the salesman sells typically have a fuel efficiency of about 35 miles per gallon on the highway.

Imagine that you now have the capacity to test 100 cars on the lot. Using the same random state of `30`, take a sample of 100 cars. Run a t-test to verify the salesman's claim with your new larger sample. Do you reach the same conclusion as you did with the sample size of 10? 

```python
### YOUR WORK HERE
# Sample 100 cars
new_sample = df.sample(100,random_state=30)
new_sample
```

###Null Hypothesis:

The cars the salesman sells typically have a fuel efficiency of about 35 miles per gallon on the highway.

###Alternative Hypothesis:

The cars the salesman sells **do not** typically have a fuel efficiency of about 35 miles per gallon on the highway.

###Confidence Level: 95%

```python
stats.ttest_1samp(new_sample['highway-mpg'],35,nan_policy='omit')
```

###Conclusion:

Based on a t-statistic of -6.52, and a p-value of 0.0000000036, I (**reject**/fail to reject) the null hypothesis that the cars the salesman sells typically have a fuel efficiency of about 35 miles per gallon on the highway.

Why might these two t-tests using the same dataset lead to different conclusions about the salesman's claim?

It's because using a larger sample size allowed me to assess the population with greater accuracy.

```python
# Use dataframe filtering to split the dataframe into two new dataframes based
#  on party. Hold all republicans in one dataframe and all democrats in the
#  other. These will be our two different "samples."
rep = df[df['party'] == 'republican']
dem = df[df['party'] == 'democrat']
```

### 2) Perform two sample T-tests on different issues and report the results.

- Find an issue that democrats support more than republicans with p < 0.01 (significant at the 99% level).
- Find an issue that republicans support more than democrats with p < 0.01 (significant at the 99% level).
- Find an issue where the difference between republicans and democrats has p > 0.1 (Not significant at the 90% level - i.e. there may not be much of a difference the two sample means)

Please for each test that you run state your null and alternative hypothesis and then write a conclusion reflecting on the null and alternative hypothesis.

Remember, that two-sample t-tests will only tell us if the two groups are *different* from one another. We'll have to look at their sample means directly or use the sign on the t-statistic to know which of the two sample means is larger.

###Null Hypothesis:

Republican support for the budget bill is equal to democrat levels of support for the immigration bill.

###Alternative Hypothesis:

Republican support for the budget bill is not equal to democrat levels of support for the immigration bill.

###Confidence Level: 99%

```python
# Find an issue that democrats support more than republicans with p < 0.01
#  (significant at the 99% level).
stats.ttest_ind(dem['budget'],rep['budget'],nan_policy='omit')
```

###Conclusion:

Based on a t-statistic of 23.213, and a p-value of 0.000...0027, I (**reject**/fail to reject) the null hypothesis that republican levels of support are equal to democrat levels of support for the budget bill. Also, based on a t-statistic of 23.213, the democrat party supports this bill far more.

###Null Hypothesis:

Republican support for the religious-groups bill is equal to democrat levels of support for the religious-groups bill.

###Alternative Hypothesis:

Republican support for the religious-groups bill is not equal to democrat levels of support for the religious-groups bill.

###Confidence Level: 99%

```python
# Find an issue that republicans support more than democrats with p < 0.01
#  (significant at the 99% level).
stats.ttest_ind(dem['south-africa'], rep['south-africa'], nan_policy='omit')
```

###Conclusion:

Based on a t-statistic of -9.74, and a p-value of 0.000...0024, I (**reject**/fail to reject) the null hypothesis that republican levels of support are equal to democrat levels of support for the religious-groups bill. Also based on a t-statistic of -1.74, the republicans support the religious-groups bill more.

###Null Hypothesis:

Republican levels of support for the water-project bill are equal to democrat levels of support for the water-project bill.

###Alternative Hypothesis:

Republican levels of support for the water-project bill are not equal to democrat levels of support for the water-project bill.

###Confidence Level: 90%

```python
# Find an issue where the difference between republicans and democrats has
#  p > 0.1 (Not significant at the 90% level - i.e. there may not be much of a
#  difference the two sample means)
stats.ttest_ind(dem['water-project'],rep['water-project'],nan_policy='omit')
```

###Conclusion:

Based on a t-statistic of -0.089, and a p-value of 0.93, I (reject/**fail to reject**) the null hypothesis that republican levels of support are equal to democrat levels of support for the water-project bill.
