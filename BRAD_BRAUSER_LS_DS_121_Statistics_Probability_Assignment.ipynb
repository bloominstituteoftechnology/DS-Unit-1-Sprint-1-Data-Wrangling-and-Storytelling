### 2) Pretend that this dataset represents the cars at a used car lot in your local town. 

- Use df.sample() to pick a random sample of 10 cars. Note that because this sample is random we are going to set the `random_state` so that all of us in the class get the same random sample. Please set your random state to `30` when using `df.sample()`

Is your sample reflective of the population value in regards to highway-mpg? Find the mean for `highway-mpg` for the entire dataset and compare your sample mean estimate to that value. Can you say that they are different? 

```python
### YOUR WORK HERE

df_sample = df.sample(10, random_state=30)

mpg = df_sample['Highway MPG']

print(mpg.mean())
print(df['Highway MPG'].mean())
```

_Can you say that they are different?_

Yes, they are different.

The salesman says the cars he sells typically have a fuel efficiency of about 35 miles per gallon on the highway. You want to verify his claim but can only test 10 cars. Using your sample of 10, test his claim and report your results.

```python
### YOUR WORK HERE

stats.ttest_1samp(mpg, 35)
```

From the sample data, With a t-statistic of 0.95 and a p-value of 0.363, I can conclude that the salesman's claim that the cars he sells have a fuel efficiency of about 35 miles is true.

Imagine that you now have the capacity to test 100 cars on the lot. Using the same random state of `30`, what would you conclude when sampling 100 cars?

```python
### YOUR WORK HERE

df_sample_2 = df.sample(100, random_state=30)

mpg_2 = df_sample_2['Highway MPG']

stats.ttest_1samp(mpg_2, 35)
```

This 100-car sample, with a t-statistic of -6.52 and a p-value of 0, has shown that the salesman's claims of selling cars with around 35 highway mpg are actually false.

*Why might these two t-tests using the same dataset lead to different conclusions about the salesman's claim?*

The bigger the data size, the more precise the mean. A data set of only 10 cars produces mean that is not very refelctive of the population, but being able to test 100 cars will be substantially more representative.

### 2) Perform two sample T-tests on different issues and report the results.

- Find an issue that democrats support more than republicans with p < 0.01 (significant at the 99% level).
- Find an issue that republicans support more than democrats with p < 0.01 (significant at the 99% level).
- Find an issue where the difference between republicans and democrats has p > 0.1 (Not significant at the 90% level - i.e. there may not be much of a difference the two sample means)

Please for each test that you run state your null and alternative hypothesis and then write a conclusion reflecting on the null and alternative hypothesis.

Remember, that two-sample t-tests will only tell us if the two groups are *different* from one another. We'll have to look at their sample means directly or use the sign on the t-statistic to know which of the two sample means is larger. 

## An issue Democrats support more than Republicans - Adoption of the Budget Resolution

**Null Hypothesis:** Democrats support adopting the budget resolution more than Republicans.

**Alternative Hypothesis:** Democrats support adopting the budget resolution less than Republicans.

**Confidence Level:** 99%

```python
stats.ttest_ind(dem_vote['Adoption of the Budget Resolution'], rep_vote['Adoption of the Budget Resolution'], nan_policy='omit')
```

**Conclusion:** With a t-statistic of -23.21 and a p-value of 0, the data suggests that Democrats supported adopting the budget resolution more than Republicans, which makes sense because the Democrats controlled the House in 1984.

## An issue Republican support more than Democrats - Crime

**Null Hypothesis:** Republicans support the crime bill more than Democrats

**Alternative Hypothesis:** Republicans support the crime bill less than Democrats

**Confidence Level:** 99%

```python
stats.ttest_ind(dem_vote['Crime'], rep_vote['Crime'], nan_policy='omit')
```

**Conclusion:** With a t-statistic of 16.34 and a p-value of 0, the data would suggest that Republicans support the crime bill more than Democrats.

## An issue that seems to have bipartisan support - Water Project Cost Sharing

**Null Hypothesis:** Democrats and Republicans give roughly equal support to Water Project Cost Sharing.

**Alternative Hypothesis:** Democrats and Republicans give different levels of support. 

**Confidence Level:** 99%

```python
stats.ttest_ind(dem_vote['Water Project Cost Sharing'], rep_vote['Water Project Cost Sharing'], nan_policy='omit')
```

With a t-statistic of 0.09 and a p-value of 0.93, the Water Project Cost Sharing bill seems to have bipartisan support.
