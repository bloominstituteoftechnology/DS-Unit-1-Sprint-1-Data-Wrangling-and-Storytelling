### Convert `int_rate`

Define a function to remove percent signs from strings and convert to floats

```python
df['int_rate'][:10].iloc[0]
```

```python
int_rate=' 14.47%'
```

```python
def int_rate_to_float(int_rate_string):
    return float(int_rate_string.strip().strip("%"))

int_rate_to_float(int_rate)
```

**CONVERT THE "TERM" COLUMN FROM STING TO INTEGER**

**MAKE A COLUMN NAMED "LOAN STATUS IS GREAT**

```python
def loan_convert(loan):
  if loan in ['Current', 'Fully Paid']:
    return '1'
  else:
    return '0'
```

```python
df1['loan_status_is_great'] = df1['loan_status'].apply(loan_convert)

df1['loan_status_is_great'].head()
```
