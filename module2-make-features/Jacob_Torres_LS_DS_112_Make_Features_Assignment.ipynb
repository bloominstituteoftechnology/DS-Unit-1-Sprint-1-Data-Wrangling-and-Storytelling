```python
# Write a function to clean the column
# which could be used for all columns with percent strings
def clean_percents(val):
  if isinstance(val, str):
    return float(val.strip().strip('%'))
  else:
    return val

# Clean the revol_util column
df['revol_util'] = df['revol_util'].apply(clean_percents)
df['revol_util'].head()
```

### Replacing uncommon job titles with "Other"

```python
# Print the top 50 job titles to see if any can be condensed
df['emp_title'].value_counts()[:50]
```

```python
# Write a function to condense duplicate or near-duplicate titles
def condense_titles(title, old_title, new_title):
  if title == old_title:
    return new_title
  else:
    return title
    
# Apply the function to duplicate nursing titles
emp_title_col = df['emp_title']
emp_title_col = emp_title_col.apply(condense_titles, args=('Rn', 'Registered Nurse'))
emp_title_col = emp_title_col.apply(condense_titles, args=('Nurse', 'Registered Nurse'))

# Check the top 50 titles again
emp_title_col.value_counts()[:50]
```

```python
# Do the same with managerial titles
manager_titles = ['Office Manager', 'General Manager', 'Sales Manager', 'Store Manager',
                  'Operations Manager', 'Project Manager', 'Assistant Manager', 'Account Manager',
                  'Program Manager', 'Service Manager', 'Branch Manager']
for title in manager_titles:
  emp_title_col = emp_title_col.apply(condense_titles, args=(title, 'Manager'))

emp_title_col.value_counts()[:20]
```

```python
# Do the same with driver duplicate titles
emp_title_col = emp_title_col.apply(condense_titles, args=('Truck Driver', 'Driver'))

# Check top 20 titles
emp_title_col.value_counts()[:20]
```

```python
# Replace the remaining titles with "Other"
top_titles = emp_title_col.value_counts()[:20]
def replace_titles_with_other(title):
  if title in top_titles:
    return title
  else:
    return 'Other'

# Apply the function
emp_title_col = emp_title_col.apply(replace_titles_with_other)
 
 # Check top titles again
print(emp_title_col.value_counts())
```
