```python
# Make a column named loan_status_is_great. It should contain the integer 1 if
# loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.

df['loan_status'].isnull().sum()
```

```python
print(df['loan_status'][:10])
print()
print(df['loan_status'][-10:])
```

```python
stat0 = df['loan_status'][0]
stat1 = df['loan_status'][1]
stat3 = df['loan_status'][128407]

print(stat0)
print(stat1)
stat3
```

```python
statlist = [stat0, stat1, stat3]
statlist
```

```python
def loanstat(obj):
    if obj == 'Current' or obj == 'Fully Paid':
      return 1
    else:
      return 0
```

```python
loanstat(statlist)
```

```python
df['loan_status_is_great'] = df['loan_status'].apply(loanstat)

df.tail()
```
