```python
# read_csv to place the csv into a dataframe, used skiprows and skipfooter
# to eliminate the weird first row and last 2 rows

df = pd.read_csv('LoanStats_2018Q4.csv', skiprows=1, skipfooter=2, engine='python')

# print the shape of the dataframe to see the number of entries
# display the head to quickly see if anything went wrong

print(df.shape)
df.head()
```

```python
df['term'].describe
```

```python
# replicating the dr.drop function to get rid of all the privacy-concerned
# information in the dataset

df = df.drop(['url', 'member_id', 'desc', 'id'], axis=1)

df.head()
```

```python
# define the int_rate_to_float function to strip the leading space and
# extemporaneous % sign

def int_rate_to_float(cell_contents):
  return float(cell_contents.strip().strip('%'))
```

```python
# call the int_rate_to_float function and write it over the
# existing int_rate series

df['int_rate'] = df['int_rate'].apply(int_rate_to_float)

df.head()
```

```python
# Test the 'cleanliness' of the column

# print(df['term'][0])
# print(df['term'][1])

# df['term'].value_counts(dropna=False).head()

# Since there are only 2 outputs, we know that every entry must have a 
# leading space, but every value in 'term' is already an int... so
# let's strip the leading space ' ' and ' months' using our
# old int_rate function but changing it a bit

df['term'] = df['term'].str.strip('months').astype(int)
```

```python
# check dtype of term

df['term'].dtype
```

```python
# double checking to make sure that everything looks good

df['term']
```

```python
# create a function that returns 1 if a loan is marked as either 'fully paid'
# or 'current', then run it on df['loan_status']

def loan_is_greater(status):
  if status == 'Fully Paid' or status == 'Current':
    return 1
  else:
    return 0
df['loan_status_is_great'] = df['loan_status'].apply(loan_is_greater)

# check to make sure it didn't fuck up
df['loan_status_is_great'].head()
```

```python
# make a series called df['last_pymnt_dt'] that mimics df['last_pymnt_d']
# except as a datetime64 instead of int

df['last_pymnt_dt'] = pd.to_datetime(df['last_pymnt_d'], infer_datetime_format=True)

# check to make sure that it looks good

df['last_pymnt_dt']
```

```python
# create the 2 columns pointing to our new payment variable

df['last_pymnt_d_month'] = df['last_pymnt_dt'].dt.month
df['last_pymnt_d_year'] = df['last_pymnt_dt'].dt.year
```
