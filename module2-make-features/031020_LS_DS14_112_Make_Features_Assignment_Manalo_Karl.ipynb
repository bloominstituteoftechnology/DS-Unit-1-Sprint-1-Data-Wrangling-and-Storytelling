```python
#importing pandas, dropping first row and last 2 rows

import pandas as pd

df = pd.read_csv('LoanStats_2018Q4.csv', skiprows=1, skipfooter=2)
df.head()
```

```python
#dropping columns with exclusively NaN values

df = df.drop(['id', 'member_id', 'url', 'desc'], axis=1)
df.head()
```

```python
#defining function to convert term from string to integer

def term_to_integer(term):
  return int(term.strip(' months'))

#applying function to dataframe via .apply() function

df['term'] = df['term'].apply(term_to_integer)
df.head()
```
