```python
import pandas as pd
pd.options.display.max_rows = 999
pd.options.display.max_columns = 999

df = pd.read_csv('LoanStats_2018Q4.csv', skiprows=1, skipfooter=2, engine='python')

# Data cleaning
df = df.drop(['url', 'member_id', 'desc', 'id'], axis=1)
print(df.shape)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.dtypes[:20]

# Things that need to be fixed:
# term - str to int
# int_rate - str to float
# emp_length - str to int
# issue_d - str to date
# earliest_cr_line - str to date
```

### term conversion

```python
df['term'] = df['term'].apply(lambda x: int(x.strip().strip(' months')))

# I'm using Lambda here because I don't want to make single-use functions for
# each individual column.
```

### int_rate conversion

```python
df['int_rate'] = df['int_rate'].apply(lambda x: float(x.strip().strip('%'))/100)
```

### emp_length conversion

```python
df['emp_length'] = df['emp_length'].fillna('0')
df['emp_length'] = df['emp_length'].apply(lambda x: int(x.strip(' years').strip('+').strip('< ')))

# This one was definitely more complicated by the Nan's, which I didn't notice
# until I accidentally printed out the column longer than I'd intended. Plus,
# they all need to be the same type so that they can be manipulated without
# making errors, then be converted back into integers.
```

### Date conversion for:
issue_d, last_pymnt_d, next_pymnt_d, and last_credit_pull_d, earliest_cr_line conversion

```python
list_of_columns = ['issue_d', 'last_pymnt_d', 'next_pymnt_d', 'last_credit_pull_d', 'earliest_cr_line']

for column in list_of_columns:
  df[column] = pd.to_datetime(df[column], infer_datetime_format=True)

# Decided to just go through all of the dates in one go instead of by hand
# since they're all the same anyway.

print(df.dtypes['issue_d'],
      df.dtypes['last_pymnt_d'],
      df.dtypes['next_pymnt_d'],
      df.dtypes['last_credit_pull_d'],
      df.dtypes['earliest_cr_line'])
```

### Column Addition
Make a column named `loan_status_is_great`. It should contain the integer 1 if `loan_status` is "Current" or "Fully Paid." Else it should contain the integer 0.

```python
df['loan_status_is_great'] = df['loan_status']
position = -1

list_of_things = []

for status in df['loan_status']:
  position += 1

  if (status == "Current") or (status == "Fully Paid"):
    list_of_things.append(1)
  else:
    list_of_things.append(0)

df['loan_status_is_great'] = list_of_things

print(df['loan_status'][20000:20010])
print(df['loan_status_is_great'][20000:20010])

# This one was a pain. I'd thought I could just overwrite any of the values
# like elsewhere in Python, but it's very finicy about it and I was having
# trouble finding out why. It's making a temporary copy that works only
# inside the for loop, so instead I made a list which overwrites the entire
# column.
```

### Further column additions
Make `last_pymnt_d_month` and `last_pymnt_d_year` columns.

```python
df['last_pymnt_d_year'], df['last_pymnt_d_month'] = df['last_pymnt_d'].dt.year, df['last_pymnt_d'].dt.month

print(df['last_pymnt_d_month'][:10],
      df['last_pymnt_d_year'][:10])
```
