```python
#Changing the string value to float64, getting rid of the white space, and the '%'
# If you can fix one value, you can fix them all:

int_rate = ' 20.89%'
```

```python
# Writing a function to fix the emp_title column

def clean_emp_title(emp_title):
  if isinstance(emp_title, str):
    return emp_title.strip().title()
  else:
    return 'Unknown'
```

2. Make a column named **loan_status_is_great**:
 Value will be 1 if loan status is Current or Fully Paid
 Value will be 0 if other

---



```python
# checking for unique values in loan_status
df['loan_status'].unique()
```

```python
# creating a list of conditions:
conditions = [df['loan_status'] == 'Fully Paid', df['loan_status'] == 'Current', df['loan_status'] == 'Charged Off',
              df['loan_status'] == 'In Grace Period', df['loan_status'] == 'Late (31-120 days)', df['loan_status'] == 'Late (16-30 days)', df['loan_status'] == 'Default']
```

```python
# running np.select with a list of conditions and values, to create the new column
df['loan_status_is_great'] = np.select(conditions, values)
df.head()
```
