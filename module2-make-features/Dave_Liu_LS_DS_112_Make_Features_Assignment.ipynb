```python
#- - Convert the `term` column from string to integer.
df['term_int'] = df.term.str.extract(r'(\d+)')  #prefer to use regex when possilble. testing by adding column first
df.drop(labels='term_int', axis=1 )             #get rid of the test column
df['term'] = df.term.str.extract(r'(\d+)')      #replace term  with the modified data
```

```python
#- Make a column named `loan_status_is_great`. It should contain the integer 1 if `loan_status` is "Current" or "Fully Paid." Else it should contain the integer 0.
df.loan_status.value_counts   
def check(status):            #heres the simple function that we can use to generate the column values
  return 1 if (status == 'Current')  or (status == 'Fully Paid') else 0
df['loan_status_is_great'] = df.loan_status.apply(check)                  #append column 
# df.head()
```
