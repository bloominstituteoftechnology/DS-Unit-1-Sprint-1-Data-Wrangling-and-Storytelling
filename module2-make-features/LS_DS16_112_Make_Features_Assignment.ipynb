```python
# Read in the csv with skip rows parameter and footer
df = pd.read_csv('LoanStats_2018Q4.csv', header=1, skipfooter=2, engine="python")

print(df.shape)
df.head()
```

```python
# Sum null values by column and sort acesding
df.isnull().sum().sort_values()
```

```python
# dropping column that contain all NaN
df = df.drop(['url', 'member_id', 'id', 'desc'], axis = 1)
df.head()
```

```python
df['int_rate'].head(10)
```

```python
# Method change str -> float
def int_rate_to_float(int_rate):
  return float(int_rate.strip().strip('%'))
```

```python
print(int_rate_to_float(' 18.94%'))
print(type(int_rate_to_float(' 18.94%')))
```

```python
# Modified int_rate column to float using apply()
df['int_rate'] = df['int_rate'].apply(int_rate_to_float)

# inspect int_rate column
df['int_rate'].head()
```

```python
# Looka t the top 20 employment title
df['emp_title'].value_counts().head(20)
```

```python
# .unique for unique value
df['emp_title'].nunique()
```

```python
# Finding NaN value
df['emp_title'].isnull().sum()
# or
df['emp_title'].value_counts(dropna=False).head(20)
```

```python
# Cleaning NaN in emp_title col
def clean_emp_title(title):
  if isinstance(title, str):
    return title.strip().title()
  else:
    return "Unknown"

df['emp_title'] = df['emp_title'].apply(clean_emp_title)
```

```python
# Checking if nan values are dropped
df['emp_title'].isnull().sum()
```

```python
# using list comprehension
[col for col in df if col.endswith('_d')]
```

```python
# Look at issue_d type
df['issue_d'].head()
```

```python
# convert issue_d to date/time
df['issue_d'] = pd.to_datetime(df['issue_d'], infer_datetime_format=True)

# Check if it's converted
df['issue_d'][0]
```

```python
df['issue_d'].head()
```

```python
# make issue month / issue year col
df['issue_year'] = df['issue_d'].dt.year
df['issue_month'] = df['issue_d'].dt.month

df.head()
```

```python
df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'], infer_datetime_format=True)
df['earliest_cr_line'].head()
```

```python
# Create new col 
df['years_from_earliest_credit_to_issue'] = ((df['issue_d'] - df['earliest_cr_line']).dt.days) / 365
df.head()
```

```python
df['years_from_earliest_credit_to_issue'].describe()
```

```python
## Convert 'term' from string to int
df['term'].head()
```

```python
def terms_to_int(term):
  return int(term.strip(' months').strip())

df['term'] = df['term'].apply(terms_to_int)
df['term'].head()
```

```python
## Making loan_status_is_great
df['loan_status'].value_counts()
```

```python
def check_loan_status(status):
  if status == "Current" or status == "Fully Paid":
    return 1
  else:
    return 0

df['loan_status_is_great'] = df['loan_status'].apply(check_loan_status)

df['loan_status_is_great'].value_counts()
```

```python
## Make last_payment_d_month
df['last_pymnt_d_month'] = pd.to_datetime(df['last_pymnt_d']).dt.month
df['last_pymnt_d_year'] = pd.to_datetime(df['last_pymnt_d']).dt.year

# checking null because value is being displayed as float
print(df['last_pymnt_d_month'].isnull().sum()) 
print(df['last_pymnt_d_year'].isnull().sum()) 
```

```python
#  convert NaN to 0 
def clear_nan(inp):
  if np.isnan(inp):
    return 0
  return int(inp)

# Clearing nan, converting float -> int
df['last_pymnt_d_month'] = df['last_pymnt_d_month'].apply(clear_nan)
df['last_pymnt_d_year'] = df['last_pymnt_d_year'].apply(clear_nan)

# Checking if nan got converted to 0
print(df['last_pymnt_d_month'].value_counts())
print(df['last_pymnt_d_year'].value_counts())
```

```python
# Final check, isnull().sum() should return 0
print(df['last_pymnt_d_month'].isnull().sum())
print(df['last_pymnt_d_year'].isnull().sum())
```
