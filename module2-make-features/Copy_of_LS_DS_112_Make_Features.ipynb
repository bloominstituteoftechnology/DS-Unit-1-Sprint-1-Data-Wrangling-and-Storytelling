## Work with strings

For machine learning, we usually want to replace strings with numbers.

We can get info about which columns have a datatype of "object" (strings)

```python
df.isnull().sum().sort_values() # we are missing in every

#why do we have at least 2 missing?
#at the tail, we have footer that is only 2 single lines screwing everything up

#the last 4 missing for every value is because to protect privacy, they let us know how many IDs they have, but don't give out. this is SOP w datasets
```

```python
df1 = df.drop(['url','member_id','desc','id'], axis=1)
df1.head()
```


### Convert `int_rate`

Define a function to remove percent signs from strings and convert to floats

```python
df['int_rate'][:10]
```

```python
# has a leading space
i_rate = ' 14.47%'
#removes whitespace at beg/end by def, can chain
float(i_rate.strip().strip('%'))

def int_rate_to_float(ir_string):
  return(float(ir_string.strip().strip('%'))
)
  
type(int_rate_to_float(i_rate))
```

Apply the function to the `int_rate` column

```python
#apply does for loop for us

df1 = df1['int_rate'].apply(int_rate_to_float)

df1.head(20)
```

### Clean `emp_title`

Look at top 20 titles

```python
df['emp_title'].value_counts(dropna=False)[:20] #includes NaN with dropna

#we want to decrease duplication; RN = registered nurse, manager = Manager for example
```

```python
len(df['emp_title'].value_counts())
```

How often is `emp_title` null?

```python
import numpy as np 
type(np.NaN)
#this means our column has strings and floats which is vaguely bad 
```

Clean the title and handle missing values

```python
examples = ["owner", 'Supervisor', 'Project Manager', 'manager', np.NaN]
```

```python
def clean_title(title):
  if isinstance(title, str):
    return title.strip().title()
  else:
    return "Unknown"

[clean_title(item) for item in examples]
```

```python
df = df['emp_title'].apply(clean_title)
```

```python
df['emp_title'].value_counts(dropna=False)[:20]
```
