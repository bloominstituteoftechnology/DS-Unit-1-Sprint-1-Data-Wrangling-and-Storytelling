```python
# Casting the string value to numeric data type, which is good for ML. 
# In this case, we'll be going from string -> float given the decimal

float(int_rate_value.strip().strip('%'))

# We know it's a float because the single quotes went adios
```

```python
# Write a function that can do what we have written above to any 
# string that is passsed to it.

# I have to copy/paste this one directly and work it out conceptually.

def int_rate_to_float(cell_contents):
  return float(cell_contents.strip().strip('%'))

int_rate_to_float(int_rate_value)

# Okay, so we define a function. "Int_rate to float" is the function name.
# (cell_contents) is the x value that each item in a list passes through
# during each iteration. The return value is what we want returned.
# We combined hte calculation part with the return part since we
# don't have a For Loop. 

# Then, when the function is called, we need the function name. Then, we
# pass in our "real world values" (not an x) to see what happens when
# our real world example, in this case, int_rate_value is passed
# throught this function

# This is essentially useless though as it does not iterate through an 
# entire column or list, which would be way more helpful. 
```

```python
# For Loop Method:

int_rate_list = []

for x in df_loans['int_rate']:
    int_rate_list.append(int_rate_to_float(x))


# Turning into a new column:

df_loans['df_int_rate_cleaned'] = pd.Series(int_rate_list)

# So, basically, we:
# 1) Defined a function that can be used on any 1 cell
# 2) Created an empty list that will eventually become a new column
# 3) Made a for loop that said "For every item in this column
# a) Do the int_rate_to_float function, then
# b) Add it tot he empty list
# 4) Created a new column called "df_int_rate_cleaned"that shows the numeric values
# using pd.Series mathod. Not sure exactly what it does, but passing the 
# int_rate_list through it made sense. 
```
