These are the problems that we need to address with this column:

-the string column that should be numeric
' 12.98%'

-Percent sign % included with the number needs to be removed

-Leading space at the beginning of the string needs to be removed.

```python
#Lets start with just fixing a single string. 
#If we can fix one, we can usually fix all of them

int_rate = ' 12.98%'
```

```python
#Removes the leading space in the string. 
#Not the string with the ' '

int_rate.strip()
```

```python
#Removes the '%' sign in the string

int_rate.strip('%')
```

```python
#The proccess of added two dot methods is called "chaining"
int_rate.strip().strip('%')
```

```python
#Let's take a second to see what type our column is.
type(int_rate.strip().strip('%'))
```

```python
#"Cast" the string value to a float.
#"Casting" 

float(int_rate.strip().strip('%'))
```

```python
#Let's check the tyoe again...

type(float(int_rate.strip().strip('%')))
```

Let's write a function to make our solution reusable for other cells in a column.

```python
# Write a function that can do what we have written above to any
# string that is passed to it.

def int_rate_to_float(cell_contents):
  return float(cell_contents.strip().strip('%'))
```

```python
#Test out the function by calling it on our example. 
# Let's Try.

int_rate_to_float(int_rate)
```

```python
#Is the data type correct. Verify.

type(int_rate_to_float(int_rate))
```

```python
cleaned = []

for int_rate in df['int_rate']:
  cleaned.append(int_rate_to_float(int_rate))

  df['int_rate_cleaned'] = pd.Series(cleaned)
```

#Create a function to apply the rule:
> Under the loan_status column, is the status is "Current" or "Fully Paid" the data needs to reflect that as the number 1. 

>Otherwise, the data in that same column needs to return with the number 0.

>After creating the function you will need to apply that rule to every cell within that column ('loan_status')
>You will need to use the .apply() function. We can reuse that function on those individual cells of a dataframe column via the .apply() function.

*Be aware of when and when not to use the quotes* **bold text**

```python
def change_status(loan_status):
  if loan_status == 'Current' or loan_status == 'Fully Paid':
    return 1
  else:
    return 0

df['loan_status'] = df['loan_status'].apply(change_status)
```
