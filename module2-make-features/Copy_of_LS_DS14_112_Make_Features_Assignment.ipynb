```python
# I'll create a function to remove it

def term_cleanup (cell):
  return cell.strip()

a = ' 60 months'
term_cleanup(a)
```

```python
# And apply the function

df['term'] = df['term'].apply(term_cleanup)

df['term'][27]
```

```python
# Now to change to integer ...
# Used .astype() function to achieve, had to research proper syntax and 
# discovered != serves as a logical operator to replace what some if statements
# might have done. I'm still a bit cloudy though and could use some help on 
# how to pull up/find the documentation for that sort of thing - I read the
# documentation on .astype() but it didn't say anything about the paremeters ...
# So in other words, I don't completely understand how this != is doing its
# thing ...

df['term'] = (df['term'] !='60 months').astype(int)
df.head()
```

```python
df['term'].head(10)

# Confirmed term values now changed to integers
```

```python
# Make a column named loan_status_is_great. It should contain the integer 1 if 
# loan_status is "Current" or "Fully Paid." Else it should contain the integer 0

def loan_type (value):
  if value == 'Current' or value == 'Fully Paid':
    return 1
  else:
    return 0
    
df['loan_status_is_great'] = df['loan_status'].apply(loan_type)

df['loan_status_is_great'].head()
```

```python
# Confirmed column is added and looks appropriate

df_loans = df[['loan_status','loan_status_is_great']].copy()

df_loans.head(50)
```

```python
# Make last_pymnt_d_month and last_pymnt_d_year columns

# Start with changing 'last_pymnt_d' to datetime

df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], infer_datetime_format=True)

df['last_pymnt_d'].describe
```
