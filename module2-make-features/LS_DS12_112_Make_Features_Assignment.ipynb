```python
### FUNCTION TIME ###
```

```python
# Create function
def int_rate_to_float(cell_contents):
  return float(cell_contents.strip().strip('%'))
```

```python
# Call the single one we made above
int_rate_to_float(int_rate)
```

```python
# Function worked if type is float
type(int_rate_to_float(int_rate))
```

```python
# Now, use the function on every cell in the row 'int_rate'
```

```python
# Make a list for vaues to go into
new_list = []
```

```python
# For loop to add to the list we just made
for cell in df['int_rate']:
  new_list.append(int_rate_to_float(cell))
```

```python
# View first 10
new_list[:10]
```

```python
# Start making the new column
df['int_rate_cleaned'] = pd.Series(new_list)
```

```python
# Check head
df.head(3)
```

```python
# apply() it
df['int_rate_cleaned_from_apply'] = df['int_rate'].apply(int_rate_to_float)
```

```python
# Examples of employee titles we want to clean up
examples = ['owner', 'Supervisor', 'Project Manager', np.NaN]
```

```python
# Function to clean them up
def clean_title(job_title):
  if isinstance(job_title, str):
    return job_title.title().strip()
  else:
    return "Unknown"
```

```python
# Show it
for title in examples:
  print(clean_title(title))
```

```python
# One line it back to a list
[clean_title(title) for title in examples]
```

```python
# Overwrite original column
df['emp_title'] = df['emp_title'].apply(clean_title)
```
