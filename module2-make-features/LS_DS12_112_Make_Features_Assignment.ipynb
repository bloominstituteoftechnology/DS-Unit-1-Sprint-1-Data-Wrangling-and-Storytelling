```python
import pandas as pd
```

```python
df = pd.read_csv('LoanStats_2018Q4.csv', header=1, skipfooter=2, engine='python')
print(df.shape)
df.head()
```

```python
pd.set_option('display.max_rows', 200)
df.isnull().sum().sort_values()
```

```python
df.shape
```

```python
df = df.drop(['url', 'member_id', 'desc', 'id'], axis=1)
```

```python
df.head()
```

```python
def int_rate_to_float(cell_contents):
  return float(cell_contents.strip().strip('%'))
```

```python
df['int_rate_cleaned'] = df['int_rate'].apply(int_rate_to_float)
df.head()
```

```python
df.dtypes[-5:]
```

```python
import numpy as np
```

```python
examples = ['owner', 'Supervisor', ' Project Manager', np.NaN]
```

```python
def clean_title(job_title):
  if isinstance(job_title, str):
    return job_title.title().strip()
  else:
    return "Unknown"
```

```python
[clean_title(title) for title in examples]
```

```python
df['emp_title'] = df['emp_title'].apply(clean_title)
df.head()
```

##Convert the term column from string to integer.

```python
term = ' 36 months'
```

```python
term.strip().strip(' months')
```

```python
int(term.strip().strip(' months'))
```

```python
type(int(term.strip().strip(' months')))
```

```python
def term_to_int(cell_contents):
  return int(cell_contents.strip().strip(' months'))
```

```python
term_to_int(term)
```

```python
df['term'] = df['term'].apply(term_to_int)
```

```python
df.head()
```

##Make a column named loan_status_is_great. It should contain the integer 1 if loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.

```python
df['loan_status'].isnull().sum()
```

```python
def str_to_int(loan_status):
  if loan_status == "Current" or loan_status == "Fully Paid":
    return 1
  else:
    return 0
```

```python
df['loan_status_is_great'] = df['loan_status'].apply(str_to_int)
```

```python
df.head()
```

##Make last_pymnt_d_month and last_pymnt_d_year columns
