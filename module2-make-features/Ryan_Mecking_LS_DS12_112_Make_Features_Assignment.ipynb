```python
#This line of code excludes the first line of the table and the last two lines to elimate the NaNs.  We specified the engine to get rid of some info displayed above the table.
df = pd.read_csv('LoanStats_2018Q4.csv', header=1, skipfooter=2, engine='python')
print(df.shape)
df.head()
```

```python
#dropped all NaN valued columns 
df = df.drop(['url', 'member_id', 'desc', 'id'], axis=1)
df.head()
```

```python
#####Convert the term column from string to integer.#####
#Going to look at the column 'term'
df['term'].head()
```

```python
#I need to elimanate the spacing in the front as well as the months after to have it read as an Integer.
term = '     36 months'
term.strip().strip(' months')
```

```python
print(int(term.strip().strip(' months')))
type(int(term.strip().strip(' months')))
```

```python
#Now that I know it works I need make it reusable. 
def term_to_int(cell_contents):
  return int(cell_contents.strip().strip(' months'))
term_to_int(term)
```

```python
#Now I need to apply the code to all the cells.
new_list = []
for cell in df['term']:
  new_list.append(term_to_int(cell))
new_list[0:10]
```

```python
#Now I will apply term_months into the table
df['term_months'] = pd.Series(new_list)
df.head()
```

```python
#making sure the new colmn is still an int.
df.dtypes[-2:]
```

```python
#doing the same thing as above but with the .apply() function
df['term_months_from_apply'] = df['term'].apply(term_to_int)
df.head()
```

```python
#####Make a column named loan_status_is_great. It should contain the integer 1 if loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.#####
def loan_status_is_great(loan_status):
  if loan_status == "Current" or loan_status == "Fully Paid":
    return 1
  else:
    return 0

df['loan_status'] = df['loan_status'].apply(loan_status_is_great)
df.head()
#cant seem to grab the two variables 
df.loan_status
```
