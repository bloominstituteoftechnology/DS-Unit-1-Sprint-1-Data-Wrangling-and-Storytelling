```python
# Read CSV
# Clean data set a little

df = pd.read_csv('LoanStats_2018Q4.csv', skiprows= 1, skipfooter= 2, engine= 'python')

df = df.drop(['url', 'member_id', 'desc', 'id'], axis=1)

print(df.shape)
df.head()
```

# Convert 'term' column to integer

```python
# no null values for term column

df['term'].isnull().sum()
```

```python
# Convert term column to int
df['term'][0]
```

```python
# has a leading space
# take off months
# convert to int


# Test
test = df['term'][0]
int(test.strip().strip(' months'))

```

```python
# Write a function

def clean_term(cell_contents):
  return int(cell_contents.strip().strip(' months'))

# Test 

clean_term(test)
```

```python
# Apply to column, and rewrite

df['term'] = df['term'].apply(clean_term)

df.head()
```

# Make a column named 'loan_status_is_great'

```python
def convert_column(cell_contents):
  if cell_contents == 'Current':
    return 1
  elif cell_contents == 'Fully Paid':
    return 1
  else:
    return 0
```

```python
# Test
test = df['loan_status'][0]
convert_column(test)
```

```python
df['loan_status_is_great'] = df['loan_status'].apply(convert_column) 

```

```python
df.head()
```

# Make new columns for last payment month and year

```python
# Infer datatime

df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], infer_datetime_format=True)
```

```python
# Make month and year columns

df['last_pymnt_d_month'] = df['last_pymnt_d'].dt.month
df['last_pymnt_d_year'] = df['last_pymnt_d'].dt.year

df.head()
```
