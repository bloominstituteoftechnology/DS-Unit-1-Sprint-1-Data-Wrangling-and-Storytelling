```python
# testing chaining of functions

term = '  36 months'

print(int(term.strip().strip('months')))

# Changing the string into an integer and running type

type(int(term.strip().strip('months')))
```

```python
# Define function to replace strings to integers

def change_str_to_int(cells_to_replace):
  return int(cells_to_replace.strip().strip('months'))
  
```

```python
# applying() the function to the column labeled 'terms'

df['term'].apply(change_str_to_int)
```

```python
df['int_term'] = df['term'].apply(change_str_to_int)

df['int_term'].head()

# replaced the original column strings with integers

df['term'] = df['term'].apply(change_str_to_int)

df['term'].head()

```

```python
examples = ['Current', 'Fully Paid', 'Charged Off', 'Late (31-120 days)', 
            'In Grace Period', 'Late (16-30 days)', 'Default']
```

```python
# Creating a function for relabeling 

def one_or_zero(cell):
  if cell == 'Fully Paid': 
    return(1)
  elif cell == 'Current':
    return(1)
  else:
    return(0)
```

```python
for example in examples:
  print(one_or_zero(example))
```

```python
df['loan_status_is_great'] = df['loan_status'].apply(one_or_zero)

new_df = df[['loan_status_is_great', 'loan_status']]

new_df.head()
```

```python
# Making a last payment month and last payment year column
# Change the format of the last_pymnt_d
```

```python
df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], infer_datetime_format=True)

df['last_pymnt_d'].head()
```

```python
# pull year and month from the new date format

df['last_pymnt_d_month'] = df['last_pymnt_d'].dt.month
df['last_pymnt_d_year'] = df['last_pymnt_d'].dt.year
```
