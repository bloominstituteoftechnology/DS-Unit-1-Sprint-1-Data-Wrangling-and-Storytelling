```python
import pandas as pd

pd.options.display.max_rows = 999
pd.options.display.max_columns = 100 

df= pd.read_csv ('LoanStats_2018Q4.csv',header=1,skipfooter=2, engine = 'python')

df.head()
```

```python
def term_int_convert(term_str):
  return int(term_str.strip().strip('months').strip())
  
term_int_convert(term)
```

```python
df= df['term'].apply(term_int_convert)
df.head()
```

```python
def loan_status_is_great (loan):
  if (loan, "Current", "Fully Paid"):
    return "1"
  else:
    return "0"
```

```python
 df_loan = df['loan_status'].apply(loan_status_is_great)
 df_loan
```

```python
def clean_title(title):
  if isinstance(title, str):
    return title.strip().title()
  else:
    return "Unknown"
```

```python
df['emp_title'] = df['emp_title'].apply(clean_title)
```
