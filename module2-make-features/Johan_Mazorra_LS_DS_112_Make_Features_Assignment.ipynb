```python
# Now we're going to strip months out of the column 'term' values so we can easily convert it into an integer.
term = '36 months'
```

```python
new_term = []

for cell in df['term']:
  new_term.append(int(cell.strip().strip(' months')))

new_term
```

```python
def new_term_to_int(cell_contents):
  return int(cell_contents.strip().strip(' months'))

new_term_to_int(term)
```

```python
type(new_term_to_int(term))
```

```python
# Now everything is set, so now we can find the new column as only integers.
df['new_term_to_int'] = df['term'].apply(new_term_to_int)

df.head(10)
```

```python
# Next, we're going to make a new column called loan_status_is_great where only 'Current' and 'Finally Paid' will be
# displayed as 1, and everything else would appear as 0.
df['loan_status'].value_counts(dropna=False).head(20)
```

```python
# Just to make sure there is no missing values so there's less of a hassle.
df['loan_status'].isnull().sum()
```

```python
# And here is the new column.
df.loc[df['loan_status']=='Current', 'loan_status_is_great'] = 1
df.loc[df['loan_status']=='Fully Paid', 'loan_status_is_great'] = 1

df['loan_status_is_great'].fillna('0', inplace=True)
```

```python
df.head(15)
```

```python
# Finally, we're going to add 2 new columns into our dataframe which are going to be
# for the last payment of the month and of the year.
df['last_pymnt_d'].value_counts()
```

```python
# It's not yet a dtype of datetime, so we're going to use the code below to make the conversion:
df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], infer_datetime_format=True)
```

```python
# Here is confirmed that we have a dtype of datetime.
df['last_pymnt_d'].describe
```

```python
df['last_pymnt_d'][0]
```

```python
df['last_pymnt_d'].dt.month
```

```python
# Since the output of the years and months came out as floats, we're going to format them as integers so
# it doesn't look messy.
pd.options.display.float_format = '{:,.0f}'.format
df['last_pymnt_d'].dt.year
```

```python
pd.options.display.float_format = '{:,.0f}'.format
df['last_pymnt_d'].dt.month
```

```python
# And finally, we have the desired columns. That wasn't too hard.
df['last_pymnt_d_year'] = df['last_pymnt_d'].dt.year
df['last_pymnt_d_month'] = df['last_pymnt_d'].dt.month

df.head()
```
