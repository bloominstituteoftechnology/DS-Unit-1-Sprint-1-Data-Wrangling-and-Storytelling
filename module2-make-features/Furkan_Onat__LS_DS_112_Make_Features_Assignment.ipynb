# Write a function to make our solution reusable!

```python
# Write a function to work for any value/cell in a dataframe column

def int_rate_to_float(dirty_int_rate):
  return float(dirty_int_rate.strip().strip('%'))
```

```python
# Test the function
int_rate_to_float(int_rate)
```

```python
int_rate_to_float(' 15.69%')
```

```python
my_int_rate_var = ' 15.69%'
int_rate_to_float(my_int_rate_var)
```

```python
# Check the datatype

type(int_rate_to_float(int_rate))
```

```python
# Apply our solution to every cell in a column

df['int_rate'].head()
```

```python
int_rates = [' 16.57%', ' 16.52%', ' 16.58%', ' 16.41%', ' 15.47%']
```

```python
clean_int_rates = []
for int_rates in int_rates:
  clean_int_rates.append(int_rate_to_float(int_rate))

  # Takes the list and turns it into a dataframe column

clean_int_rates
```

```python
# The technical name for a column of a dataframe is a 'Series'
clean_int_rates = pd.Series(clean_int_rates)

clean_int_rates
```

# The .apply() function

```python
df['int_rate_cleaned'] = df['int_rate'].apply(int_rate_to_float)

df.head()
```

```python
# Create some examples
import numpy as np

examples = ['owner', 'Supervisor', ' Project Manager', np.NaN]
```

```python
# Write a function to clean up these use cases and increase uniformity

def clean_emp_title(title):
  if isinstance(title, str):
    return title.strip().title()
  else:
    return 'Unknown'

for example in examples:
  print(clean_emp_title(example))
```
