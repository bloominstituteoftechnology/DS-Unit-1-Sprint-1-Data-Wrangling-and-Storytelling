```python
#Use the pandas module read and put data into dataframe
import pandas as pd
df = pd.read_csv('LoanStats_2018Q4.csv', skiprows=1, skipfooter=2, engine='python')
df.head()
```

Convert term

```python
#Define mystring as a function 
def term_2_float(mystring):
  return float(mystring.strip().strip('months '))
```

```python
#Test the function
term_2_float(term_1)
```

```python
#Use the apply method to apply the function to the entire column
df['term_float'] = df['term'].apply(term_2_float)
df.columns
```

```python
#Compare the old and new columns
df[['term', 'term_float']].head()
```

```python
#Confirm that we can now employ math on this
df['term_double']=df['term_float']*2
df[['term', 'term_float', 'term_double']].sample(5)
```

Clean the title and missing values

```python
import numpy as np
type(np.NaN)
```

```python
# Write a function to clean up our titles
def clean_title(title):
  if isinstance(title, str):
    return title.strip().title()
  else:
    return 'Unknown'

examples=['owner', 'Supervisor', ' Project Manager', np.NaN]

for item in examples:
  print(clean_title(item))
```

```python
# Now that we've tested our function, let's apply it to the entire dataframe column.
df['emp_title_clean']=df['emp_title'].apply(clean_title)
df.columns
```

```python
# compare old and new values
df[['emp_title', 'emp_title_clean']].sample(6)
```
