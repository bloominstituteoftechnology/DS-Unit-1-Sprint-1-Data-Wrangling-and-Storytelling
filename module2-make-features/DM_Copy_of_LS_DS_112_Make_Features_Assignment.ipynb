### Make a column named loan_status_is_great. It should contain the integer 1 if loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.

```python
df['loan_status'].head(20)
```

```python
df['loan_status'].value_counts(dropna=False)
```

```python
stati = ['Current', 'Fully Paid', 'Charged Off', 'Late', 'Default', 'Current']
```

```python
def replace_status(stati):
    if stati == "Current":
          return 1
    elif stati == "Fully Paid":
          return 1
    else:
          return 0

for status in stati:
  print(replace_status(status))
```

```python
df['loan_status_great'] = df['loan_status'].apply(replace_status)
```

```python
df.head()
```

## Make last_pymnt_d_month and last_pymnt_d_year columns.

```python
df['last_pymnt_d'].isnull().sum()
```

```python
df['last_pymnt_d_month'] = pd.to_datetime(df['last_pymnt_d'].fillna(0), infer_datetime_format=True).dt.month
df['last_pymnt_d_year'] = pd.to_datetime(df['last_pymnt_d'].fillna(0), infer_datetime_format=True).dt.year
```
