```python
term = ' 36 months'
```

```python
# Remove leading space
term.strip()
```

```python
# Remove months from string
term.strip('months')
```

```python
# Remove leading space, months, and ending space
term.strip().strip('months').rstrip()
```

```python
# Check the value of our variable after the characters has been stripped
type(term.strip().strip('months').rstrip())
```

```python
# "Cast" the string value to an interger
int(term.strip().strip('months').rstrip())
```

```python
# Check that we have successfully "cast" the string to an interger value
type(int(term.strip().strip('months').rstrip()))
```

```python
# Write a function that can do what we have written above to any
# string that is passed to it.
def term_to_int(cell_data):
  return int(cell_data.strip().strip('months').rstrip())
```

```python
# Test out our function by calling it on our example
term_to_int(term)
```

```python
# Is the data type correct?
type(term_to_int(term))
```

```python
# Applying our solution to every cell in a column
df['term_to_int'] = df['term'].apply(term_to_int)

df.head()
```
