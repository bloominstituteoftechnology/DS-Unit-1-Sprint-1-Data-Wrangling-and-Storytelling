```python
#Make a column named loan_status_is_great. It should contain the integer 1 if loan_status is "Current" or "Fully Paid." 
#Else it should contain the integer 0.
```

```python
def loan_status_checker(loan_status):
  if loan_status == 'Current' or loan_status == 'Fully Paid':
    return 1
  else:
    return 0
```

```python
#Make last_pymnt_d_month and last_pymnt_d_year columns.
```

```python
def find_last_pay_month(last_pymnt_d):
  if last_pymnt_d == np.NaN:
    return "Unknown"
  else:
    return pd.to_datetime((last_pymnt_d), infer_datetime_format=True).month
```

```python
find_last_pay_month('Dec-2018')
```

```python
df['last_pymnt_d_month']=df['last_pymnt_d'].apply(find_last_pay_month)
df.head(55)
```

```python
def find_last_pay_year(last_pymnt_d):
  if last_pymnt_d == np.NaN:
    return "Unknown"
  else:
    return pd.to_datetime((last_pymnt_d), infer_datetime_format=True).year
```

```python
find_last_pay_year('Oct-2020')
```

```python
df['last_pymnt_d_year']=df['last_pymnt_d'].apply(find_last_pay_year)
df
```
