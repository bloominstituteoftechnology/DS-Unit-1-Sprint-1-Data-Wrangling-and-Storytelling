```python
#Funtion to convert percents to floats
#Function looks to see if cell contents are already a float, and if not changes the contents to a float
def percent_to_float(cell_contents):
  if isinstance(cell_contents, float):
    return (cell_contents)
  else:
    return float(cell_contents.strip().strip('%'))
```

```python
#Applypercent_to_float to interest rate column
df['int_rate'] = df['int_rate'].apply(percent_to_float)
df.head()
```

```python
#Apply percent_to_float to revolving utilization data

df['revol_util'] = df['revol_util'].apply(percent_to_float)
df.head()
```

```python
#Apply percent_to_float to secondary applicant revolving utilization data

df['sec_app_revol_util'] = df['sec_app_revol_util'].apply(percent_to_float)
df.head()
```

```python
#Looks for column titles that end in '_d
[col for col in df if col.endswith('_d')]
```

```python
#Convert the term column from a string to an integer
df['term'].head()
```

```python
#The values have the word 'month' in them, so, we have to remove it.
test = '36 months'
int(test.strip('months').strip())
```

```python
#Okay, now to make a function
def term_clean(cell_contents):
  return int(cell_contents.strip('months').strip())
```

```python
#...And apply it to the term column
df['term'] = df['term'].apply(term_clean)
df['term'].head()
```

```python
#Now, the loan status is great column
#First, check the data type in the loan status column

df['loan_status'].value_counts(dropna=False)
```

```python
#I'm going to change the goal to include 'issued' and 'in grace period' because those aren't delinquent yet
examples = ['Current', 'Issued', 
            'Fully Paid', 'In Grace Period', 
            'Late (31-120 days)', 'Late (16-30 days)']
```

```python
def great_loan(status):
  if status == 'Current':
    return 1
  elif status =='Issued':
    return 1
  elif status =='Fully Paid':
    return 1
  elif status =='In Grace Period':
    return 1
  else:
    return 0
```

```python
great_loans = []
for example in examples:
  great_loans.append(great_loan(example))

great_loans
```

```python
#Yay! That worked! Now to create the new column and apply the function.
df['loan_status_is_great'] = df['loan_status'].apply(great_loan)
df.head()
```

Make last_payment_d_year and last_payment_d_month columns

```python
#Make the last payment year and month columns
df['last_pymnt_d_year'] = df['last_pymnt_d'].dt.year 
df['last_pymnt_d_month'] = df['last_pymnt_d'].dt.month 
df.head()
```
