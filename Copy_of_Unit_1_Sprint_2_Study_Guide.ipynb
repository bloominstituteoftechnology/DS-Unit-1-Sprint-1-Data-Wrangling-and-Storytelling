1. What is a CSV and how do you load it into a notebook?
```
A CSV file is a list of data separated by commas. To upload it in a notebook (python) you use the import csv command of any file; for example first import it using the import panda as pd; follow by using data = pd.read_csv("filename.csv") command.

```

2. What is a NaN and what will you do with it?
```
A NAN is a non numerical value, the acronym means Not A Number.We can detect, remove, and replace null values in Pandas DataFrame using these commands; isnull(), notnull(), dropna(), fillna(), replace(), interpolate()
```

3. How do you check for datatypes?
```
To check for the entire data type you use "df.dtypes" command. To check the data type of a specific column you use "df['DataFrame Column'].dtypes" command.

4. What is Crosstabs and what are they used for?
```
Cross tabs are use for summarizing data of two or more factors. Cross Tabulation is used to aggregate and jointly display the distribution of two or more variables by tabulating their results one against the other in 2-dimensional grids.
```
5. What is the difference between a T-Test and a P-Test?
```
The t-value is specific thing for a specific statistical test, that means little by itself. The p-value tells you the statistical significance of the difference.
```
